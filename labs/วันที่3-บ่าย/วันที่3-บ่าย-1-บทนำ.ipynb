# วันที่ 3: การสร้าง Dashboard และวิเคราะห์กรณีศึกษา
## ช่วงบ่าย (13.00 น.–16.00 น.): กรณีศึกษา – วิเคราะห์ข้อมูลจากสถานการณ์จริง

**วันที่:** 30 เมษายน 2568  
**เวลา:** 13.00 น.–16.00 น.  

## บทนำ

ในช่วงบ่ายนี้ เราจะนำความรู้จากช่วงเช้าเกี่ยวกับการสร้าง Dashboard และการเลือกกราฟที่เหมาะสมมาประยุกต์ใช้กับกรณีศึกษาจริง โดยเราจะวิเคราะห์ข้อมูลการขายของบริษัทค้าปลีกและสร้าง Dashboard ที่มีประสิทธิภาพใน Power BI

## 1. แนะนำกรณีศึกษา: บริษัท "รุ่งเรืองค้าปลีก"

### 1.1 ภาพรวมของบริษัท

บริษัท "รุ่งเรืองค้าปลีก" เป็นบริษัทค้าปลีกที่มีสาขาทั่วประเทศไทย จำหน่ายสินค้าหลากหลายประเภท ทั้งอิเล็กทรอนิกส์ เครื่องใช้ในบ้าน เครื่องแต่งกาย อาหารและเครื่องดื่ม และเครื่องสำอาง ผู้บริหารต้องการติดตามผลการดำเนินงานของบริษัทและวิเคราะห์ข้อมูลเพื่อหาโอกาสในการเพิ่มยอดขายและกำไร

### 1.2 ชุดข้อมูลที่ใช้ในการวิเคราะห์

บริษัทมีข้อมูลหลักที่เกี่ยวข้อง 4 ชุด:

1. **ข้อมูลสินค้า (products)**: รหัสสินค้า, ชื่อสินค้า, หมวดหมู่, หมวดหมู่ย่อย, ราคา, ต้นทุน
2. **ข้อมูลลูกค้า (customers)**: รหัสลูกค้า, ชื่อลูกค้า, ภูมิภาค, จังหวัด, เพศ, กลุ่มอายุ, วันที่ลงทะเบียน
3. **ข้อมูลร้านค้า (stores)**: รหัสร้าน, ชื่อร้าน, ภูมิภาค, จังหวัด, ประเภทร้าน, วันที่เปิดร้าน
4. **ข้อมูลการขาย (sales)**: รหัสธุรกรรม, วันที่ขาย, รหัสลูกค้า, รหัสร้าน, รหัสสินค้า, จำนวน, ราคาต่อหน่วย, ยอดขาย, ต้นทุน, วิธีการชำระเงิน

### 1.3 วัตถุประสงค์ของกรณีศึกษา

1. วิเคราะห์ข้อมูลการขายและหาข้อมูลเชิงลึกที่มีประโยชน์
2. สร้าง Dashboard ที่มีประสิทธิภาพใน Power BI ตามหลักการที่ได้เรียนรู้
3. นำเสนอข้อมูลเชิงลึกและข้อเสนอแนะที่มีประโยชน์สำหรับธุรกิจ
