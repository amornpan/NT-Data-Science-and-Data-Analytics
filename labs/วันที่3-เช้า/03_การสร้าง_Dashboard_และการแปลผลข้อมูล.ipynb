# การสร้าง Dashboard และการแปลผลข้อมูล

## บทนำ

การสร้าง Dashboard ที่มีประสิทธิภาพเป็นทักษะสำคัญสำหรับนักวิเคราะห์ข้อมูลและนักวิทยาศาสตร์ข้อมูล Dashboard ที่ดีช่วยให้ผู้ใช้สามารถเข้าใจข้อมูลที่ซับซ้อนได้อย่างรวดเร็ว และนำไปสู่การตัดสินใจที่ดีขึ้น บทเรียนนี้จะแนะนำหลักการออกแบบ Dashboard ที่มีประสิทธิภาพและวิธีการแปลผลข้อมูลอย่างถูกต้อง

## หลักการออกแบบ Dashboard ที่มีประสิทธิภาพ

### 1. การแบ่งโซนตามลำดับความสำคัญของข้อมูล

การจัดระเบียบข้อมูลตามความสำคัญช่วยให้ผู้ใช้เข้าถึงข้อมูลที่สำคัญที่สุดได้อย่างรวดเร็ว โดยทั่วไปแล้ว Dashboard ควรแบ่งออกเป็นโซนดังนี้:

1. **Header (Level Zero)**
   - องค์ประกอบ: ชื่อ Dashboard, ปุ่มนำทาง, ตัวกรองข้อมูล (Filters)
   - ตำแหน่ง: ด้านบนสุดของ Dashboard
   - ความสำคัญ: ช่วยให้ผู้ใช้เข้าใจว่ากำลังดูข้อมูลอะไร และสามารถปรับเปลี่ยนมุมมองได้ตามต้องการ

2. **ระดับที่ 1 - KPIs (Key Performance Indicators)**
   - องค์ประกอบ: การ์ดแสดงตัวชี้วัดหลักที่สำคัญ
   - ตำแหน่ง: ถัดจาก Header ลงมา (มักอยู่ด้านบนของพื้นที่เนื้อหาหลัก)
   - ความสำคัญ: แสดงภาพรวมของประสิทธิภาพหรือสถานะปัจจุบัน
   - ตัวอย่าง: ยอดขายรวม, จำนวนลูกค้า, อัตรากำไร

3. **ระดับที่ 2 - การแสดงผลรวมย่อยและแนวโน้ม**
   - องค์ประกอบ: กราฟแสดงแนวโน้ม, แผนภูมิเปรียบเทียบ
   - ตำแหน่ง: ส่วนกลางของ Dashboard
   - ความสำคัญ: แสดงข้อมูลเชิงลึกเกี่ยวกับแนวโน้มและการเปรียบเทียบระหว่างช่วงเวลาหรือหมวดหมู่

4. **ระดับที่ 3 - รายละเอียด**
   - องค์ประกอบ: ตาราง, แผนภูมิรายละเอียด
   - ตำแหน่ง: ส่วนล่างของ Dashboard
   - ความสำคัญ: แสดงข้อมูลเชิงลึกที่ละเอียดสำหรับการวิเคราะห์เชิงลึก

### 2. การเลือกรูปแบบการแสดงผลที่เหมาะสม

การเลือกรูปแบบการแสดงผลที่เหมาะสมกับประเภทข้อมูลช่วยให้ผู้ใช้เข้าใจข้อมูลได้ดียิ่งขึ้น

#### 2.1 การเปรียบเทียบค่า
- **แผนภูมิแท่ง (Bar Chart)**: เหมาะสำหรับเปรียบเทียบค่าระหว่างหมวดหมู่ต่างๆ
- **แผนภูมิแท่งแนวนอน (Horizontal Bar Chart)**: เหมาะสำหรับเปรียบเทียบค่าที่มีชื่อหมวดหมู่ยาว
- **ตัวอย่างการใช้งาน**: เปรียบเทียบยอดขายระหว่างหมวดหมู่สินค้า, เปรียบเทียบประสิทธิภาพระหว่างแผนก

#### 2.2 แนวโน้มตามเวลา
- **แผนภูมิเส้น (Line Chart)**: เหมาะสำหรับแสดงแนวโน้มหรือการเปลี่ยนแปลงตามเวลา
- **แผนภูมิพื้นที่ (Area Chart)**: เหมาะสำหรับแสดงปริมาณที่เปลี่ยนแปลงตามเวลาและเน้นปริมาณรวม
- **ตัวอย่างการใช้งาน**: แนวโน้มยอดขายรายเดือน, การเปลี่ยนแปลงของราคาหุ้นตามเวลา

#### 2.3 ส่วนประกอบของทั้งหมด
- **แผนภูมิวงกลม (Pie Chart)**: เหมาะสำหรับแสดงสัดส่วนของส่วนประกอบต่างๆ (ไม่ควรมีหมวดหมู่เกิน 6 หมวด)
- **แผนภูมิโดนัท (Donut Chart)**: คล้ายกับแผนภูมิวงกลมแต่มีพื้นที่ตรงกลางสำหรับแสดงข้อมูลเพิ่มเติม
- **ตัวอย่างการใช้งาน**: สัดส่วนยอดขายตามประเภทสินค้า, การจัดสรรงบประมาณ

#### 2.4 การกระจายของข้อมูล
- **แผนภูมิกระจาย (Scatter Plot)**: เหมาะสำหรับแสดงความสัมพันธ์ระหว่างตัวแปรสองตัว
- **Heatmap**: เหมาะสำหรับแสดงความสัมพันธ์ระหว่างหมวดหมู่สองกลุ่มด้วยสี
- **ตัวอย่างการใช้งาน**: ความสัมพันธ์ระหว่างราคาและปริมาณการขาย, การวิเคราะห์ความหนาแน่นของลูกค้าตามพื้นที่

#### 2.5 ข้อมูลภูมิศาสตร์
- **แผนที่ (Map)**: เหมาะสำหรับแสดงข้อมูลที่มีความเกี่ยวข้องกับตำแหน่งทางภูมิศาสตร์
- **ตัวอย่างการใช้งาน**: ยอดขายตามภูมิภาค, การกระจายตัวของลูกค้า

### แผนภาพแสดงการตัดสินใจเลือกชนิดของกราฟ

ขั้นตอนการเลือกชนิดของกราฟที่เหมาะสมกับข้อมูล:

1. **ถ้าต้องการเปรียบเทียบค่าระหว่างหมวดหมู่**
   - ถ้ามีจำนวนหมวดหมู่น้อย → ใช้แผนภูมิแท่ง (Bar Chart)
   - ถ้ามีจำนวนหมวดหมู่มาก → ใช้แผนภูมิแท่งแนวนอน (Horizontal Bar Chart)

2. **ถ้าต้องการแสดงแนวโน้มตามเวลา**
   - ถ้าต้องการเน้นปริมาณรวม → ใช้แผนภูมิพื้นที่ (Area Chart)
   - ถ้าไม่ต้องการเน้นปริมาณรวม → ใช้แผนภูมิเส้น (Line Chart)

3. **ถ้าต้องการแสดงสัดส่วนของทั้งหมด**
   - ถ้ามีจำนวนหมวดหมู่น้อยกว่า 6 → ใช้แผนภูมิวงกลม (Pie Chart)
   - ถ้ามีจำนวนหมวดหมู่มากกว่า 6 → ใช้แผนภูมิแท่ง (Bar Chart)

4. **ถ้าข้อมูลมีพิกัดภูมิศาสตร์** → ใช้แผนที่ (Map)

5. **ถ้าต้องการศึกษาความสัมพันธ์ระหว่างตัวแปร** → ใช้แผนภูมิกระจาย (Scatter Plot)

6. **ถ้าต้องการแสดงรายละเอียดข้อมูลทั้งหมด** → ใช้ตาราง (Table)

### 3. การจัดวางองค์ประกอบให้น่าอ่าน

การจัดวางองค์ประกอบอย่างเหมาะสมช่วยให้ผู้ใช้สามารถเข้าถึงข้อมูลได้อย่างมีประสิทธิภาพ

#### 3.1 หลักการ Z-Pattern
- ผู้ใช้มักจะอ่านข้อมูลจากซ้ายไปขวา และจากบนลงล่าง เป็นรูปแบบตัว Z
- จัดวางข้อมูลสำคัญที่สุดไว้ตามเส้นทางการอ่านแบบ Z

#### 3.2 การกำหนดสัดส่วนพื้นที่
- ให้พื้นที่แก่แต่ละส่วนตามความสำคัญ
- หลีกเลี่ยงการมีพื้นที่ว่างมากเกินไปหรือแน่นเกินไป

#### 3.3 การจัดกลุ่มข้อมูลที่เกี่ยวข้องกัน
- จัดข้อมูลที่เกี่ยวข้องกันไว้ใกล้กัน
- ใช้เส้นขอบหรือพื้นหลังเพื่อแบ่งกลุ่มข้อมูล

### 4. การใช้สีและการออกแบบที่เหมาะสม

การใช้สีและการออกแบบที่เหมาะสมช่วยเพิ่มความน่าสนใจและความเข้าใจ

#### 4.1 การใช้สีอย่างมีความหมาย
- ใช้สีเพื่อสื่อความหมาย ไม่ใช่เพียงเพื่อความสวยงาม
- ใช้สีที่สอดคล้องกับความหมายที่ต้องการสื่อ (เช่น สีแดงสำหรับความเสี่ยง สีเขียวสำหรับความสำเร็จ)
- คำนึงถึงผู้ใช้ที่มีปัญหาในการแยกสี (Color blindness)

#### 4.2 การใช้ตัวอักษรและขนาด
- ใช้ฟอนต์ที่อ่านง่ายและเป็นมาตรฐาน
- กำหนดขนาดตัวอักษรตามลำดับความสำคัญ
- หลีกเลี่ยงการใช้ตัวอักษรหลายแบบ

#### 4.3 การใช้พื้นที่ว่าง (White Space)
- ใช้พื้นที่ว่างเพื่อแยกส่วนต่างๆ และลดความแออัด
- ไม่ควรบรรจุข้อมูลมากเกินไปในหน้าเดียว

## การแปลผลข้อมูลจาก Dashboard

การแปลผลข้อมูลอย่างถูกต้องเป็นทักษะสำคัญในการใช้ประโยชน์จาก Dashboard

### 1. การอ่านแนวโน้มและรูปแบบ

#### 1.1 การระบุแนวโน้ม
- **แนวโน้มเพิ่มขึ้น/ลดลง**: สังเกตทิศทางการเปลี่ยนแปลงของข้อมูลตามเวลา
- **แนวโน้มตามฤดูกาล**: สังเกตรูปแบบที่เกิดขึ้นซ้ำในช่วงเวลาเดียวกันของแต่ละปี
- **การเปลี่ยนแปลงแบบไม่สม่ำเสมอ**: สังเกตการเปลี่ยนแปลงที่ไม่เป็นไปตามแนวโน้มทั่วไป

#### 1.2 การเปรียบเทียบกับเป้าหมาย
- เปรียบเทียบผลลัพธ์ปัจจุบันกับเป้าหมายที่ตั้งไว้
- ระบุว่าผลลัพธ์อยู่เหนือหรือต่ำกว่าเป้าหมายเพียงใด

#### 1.3 การเปรียบเทียบระหว่างหมวดหมู่
- ระบุหมวดหมู่ที่มีประสิทธิภาพสูงสุดและต่ำสุด
- วิเคราะห์ความแตกต่างระหว่างหมวดหมู่

````python
# ตัวอย่างโค้ดการวิเคราะห์แนวโน้มข้อมูลยอดขายรายเดือน
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# สร้างข้อมูลตัวอย่าง
np.random.seed(42)
dates = pd.date_range(start='2023-01-01', end='2023-12-31', freq='M')
sales = np.random.normal(loc=1000, scale=200, size=len(dates))
# เพิ่มความผันผวนตามฤดูกาล
sales[5:8] += 300  # เพิ่มยอดขายในช่วงกลางปี (มิถุนายน-สิงหาคม)
sales[11] += 500   # เพิ่มยอดขายในช่วงปลายปี (ธันวาคม)

# สร้าง DataFrame
sales_data = pd.DataFrame({
    'date': dates,
    'sales': sales
})

# กำหนดเป้าหมายยอดขายรายเดือน
target = 1200

# พล็อตกราฟแนวโน้มยอดขายเทียบกับเป้าหมาย
plt.figure(figsize=(12, 6))
plt.plot(sales_data['date'], sales_data['sales'], marker='o', linestyle='-', color='blue', label='ยอดขายจริง')
plt.axhline(y=target, color='red', linestyle='--', label='เป้าหมาย')
plt.title('แนวโน้มยอดขายรายเดือนเทียบกับเป้าหมาย')
plt.xlabel('เดือน')
plt.ylabel('ยอดขาย (บาท)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# วิเคราะห์แนวโน้ม
month_names = ['ม.ค.', 'ก.พ.', 'มี.ค.', 'เม.ย.', 'พ.ค.', 'มิ.ย.', 'ก.ค.', 'ส.ค.', 'ก.ย.', 'ต.ค.', 'พ.ย.', 'ธ.ค.']
sales_data['month_name'] = [month_names[i] for i in sales_data['date'].dt.month - 1]
sales_data['diff_from_target'] = sales_data['sales'] - target
sales_data['performance'] = ['สูงกว่าเป้าหมาย' if x > 0 else 'ต่ำกว่าเป้าหมาย' for x in sales_data['diff_from_target']]

# แสดงผลการวิเคราะห์
print("การวิเคราะห์ยอดขายเทียบกับเป้าหมาย:")
print(sales_data[['month_name', 'sales', 'diff_from_target', 'performance']])
````

### 2. การระบุความผิดปกติและโอกาส

#### 2.1 การตรวจจับความผิดปกติ
- สังเกตข้อมูลที่แตกต่างจากแนวโน้มหรือรูปแบบทั่วไป
- พิจารณาว่าความผิดปกตินั้นเป็นโอกาสหรือความเสี่ยง

#### 2.2 การระบุช่องว่างและโอกาส
- ระบุพื้นที่ที่มีประสิทธิภาพต่ำกว่าที่ควรจะเป็น
- ค้นหาโอกาสในการปรับปรุงหรือเติบโต

#### 2.3 การสร้างสมมติฐาน
- ตั้งสมมติฐานเกี่ยวกับสาเหตุของรูปแบบหรือแนวโน้มที่สังเกตเห็น
- ทดสอบสมมติฐานด้วยการวิเคราะห์เพิ่มเติม

````python
# ตัวอย่างโค้ดการวิเคราะห์ความผิดปกติ
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# สร้างข้อมูลตัวอย่าง
np.random.seed(42)
dates = pd.date_range(start='2023-01-01', end='2023-12-31', freq='D')
daily_sales = np.random.normal(loc=1000, scale=100, size=len(dates))

# เพิ่มความผิดปกติบางวัน
anomaly_days = [45, 120, 210, 300]  # วันที่มีความผิดปกติ
for day in anomaly_days:
    if day % 2 == 0:
        daily_sales[day] += 500  # ความผิดปกติเชิงบวก
    else:
        daily_sales[day] -= 500  # ความผิดปกติเชิงลบ

# สร้าง DataFrame
daily_data = pd.DataFrame({
    'date': dates,
    'sales': daily_sales
})

# ตรวจจับความผิดปกติโดยใช้ Z-score
z_scores = stats.zscore(daily_data['sales'])
threshold = 3
daily_data['z_score'] = z_scores
daily_data['is_anomaly'] = abs(z_scores) > threshold

# พล็อตกราฟข้อมูลรายวันและระบุความผิดปกติ
plt.figure(figsize=(15, 6))
plt.plot(daily_data['date'], daily_data['sales'], color='blue', alpha=0.5, label='ยอดขายรายวัน')

# แสดงจุดที่เป็นความผิดปกติ
anomalies = daily_data[daily_data['is_anomaly']]
plt.scatter(anomalies['date'], anomalies['sales'], color='red', s=100, label='ความผิดปกติ')

plt.title('การตรวจจับความผิดปกติของยอดขายรายวัน')
plt.xlabel('วันที่')
plt.ylabel('ยอดขาย (บาท)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()
plt.show()

# แสดงรายละเอียดของความผิดปกติที่พบ
print("ความผิดปกติที่พบ:")
print(anomalies[['date', 'sales', 'z_score']])
````

### 5. ตรวจสอบการกรองข้อมูลร่วมกัน (cross-filtering)

````python
# ตัวอย่างโค้ดการสร้าง Dashboard แบบง่ายด้วย matplotlib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.gridspec import GridSpec

# สร้างข้อมูลตัวอย่าง
np.random.seed(42)

# ข้อมูลยอดขายรายเดือน
months = ['ม.ค.', 'ก.พ.', 'มี.ค.', 'เม.ย.', 'พ.ค.', 'มิ.ย.', 'ก.ค.', 'ส.ค.', 'ก.ย.', 'ต.ค.', 'พ.ย.', 'ธ.ค.']
monthly_sales = np.random.normal(loc=1000, scale=200, size=12)
monthly_sales[5:8] += 300  # เพิ่มยอดขายในช่วงกลางปี
monthly_sales[11] += 500   # เพิ่มยอดขายในช่วงเทศกาลปีใหม่

# ข้อมูลยอดขายตามหมวดหมู่
categories = ['อิเล็กทรอนิกส์', 'เสื้อผ้า', 'อาหาร', 'เครื่องใช้ในบ้าน', 'เครื่องสำอาง']
category_sales = np.random.normal(loc=800, scale=150, size=5)

# ข้อมูลยอดขายตามภูมิภาค
regions = ['กรุงเทพฯ', 'ภาคเหนือ', 'ภาคกลาง', 'ภาคอีสาน', 'ภาคใต้']
region_sales = np.random.normal(loc=1500, scale=300, size=5)

# กำหนดเป้าหมายและค่า KPI
total_sales = sum(monthly_sales)
average_sales = np.mean(monthly_sales)
max_month_sales = np.max(monthly_sales)
growth_rate = ((monthly_sales[-1] - monthly_sales[0]) / monthly_sales[0]) * 100

# สร้าง Dashboard
plt.figure(figsize=(15, 10))
gs = GridSpec(3, 4, figure=plt.gcf())  # แบ่งกริด 3 แถว 4 คอลัมน์

# KPI Cards (แถวที่ 1)
ax1 = plt.subplot(gs[0, 0])
ax1.text(0.5, 0.5, f"ยอดขายรวม\n{total_sales:,.0f} บาท", ha='center', va='center', fontsize=12)
ax1.axis('off')
ax1.set_facecolor('#f0f7ff')
ax1.patch.set_boxstyle('round,pad=0.5')

ax2 = plt.subplot(gs[0, 1])
ax2.text(0.5, 0.5, f"ยอดขายเฉลี่ย/เดือน\n{average_sales:,.0f} บาท", ha='center', va='center', fontsize=12)
ax2.axis('off')
ax2.set_facecolor('#f7fff0')
ax2.patch.set_boxstyle('round,pad=0.5')

ax3 = plt.subplot(gs[0, 2])
ax3.text(0.5, 0.5, f"ยอดขายสูงสุด\n{max_month_sales:,.0f} บาท", ha='center', va='center', fontsize=12)
ax3.axis('off')
ax3.set_facecolor('#fff0f7')
ax3.patch.set_boxstyle('round,pad=0.5')

ax4 = plt.subplot(gs[0, 3])
ax4.text(0.5, 0.5, f"อัตราการเติบโต\n{growth_rate:.1f}%", ha='center', va='center', fontsize=12)
ax4.axis('off')
ax4.set_facecolor('#f0f0ff')
ax4.patch.set_boxstyle('round,pad=0.5')

# แนวโน้มยอดขายรายเดือน (แถวที่ 2 คอลัมน์ 1-2)
ax5 = plt.subplot(gs[1, 0:2])
ax5.plot(months, monthly_sales, marker='o', linestyle='-', color='blue')
ax5.set_title('แนวโน้มยอดขายรายเดือน')
ax5.grid(True, linestyle='--', alpha=0.7)
ax5.set_ylabel('ยอดขาย (บาท)')
ax5.set_xticks(range(len(months)))
ax5.set_xticklabels(months, rotation=45)

# ยอดขายตามหมวดหมู่ (แถวที่ 2 คอลัมน์ 3-4)
ax6 = plt.subplot(gs[1, 2:4])
bars = ax6.bar(categories, category_sales, color='orange')
ax6.set_title('ยอดขายตามหมวดหมู่สินค้า')
ax6.set_ylabel('ยอดขาย (บาท)')
ax6.set_xticks(range(len(categories)))
ax6.set_xticklabels(categories, rotation=45)

# ยอดขายตามภูมิภาค (แถวที่ 3 คอลัมน์ 1-2)
ax7 = plt.subplot(gs[2, 0:2])
ax7.pie(region_sales, labels=regions, autopct='%1.1f%%', startangle=90)
ax7.set_title('สัดส่วนยอดขายตามภูมิภาค')
ax7.axis('equal')

# ตารางสรุป (แถวที่ 3 คอลัมน์ 3-4)
ax8 = plt.subplot(gs[2, 2:4])
ax8.axis('tight')
ax8.axis('off')
table_data = [
    ['ภูมิภาค', 'ยอดขาย', '% ของยอดขายรวม'],
    [regions[0], f"{region_sales[0]:,.0f}", f"{(region_sales[0]/sum(region_sales))*100:.1f}%"],
    [regions[1], f"{region_sales[1]:,.0f}", f"{(region_sales[1]/sum(region_sales))*100:.1f}%"],
    [regions[2], f"{region_sales[2]:,.0f}", f"{(region_sales[2]/sum(region_sales))*100:.1f}%"],
    [regions[3], f"{region_sales[3]:,.0f}", f"{(region_sales[3]/sum(region_sales))*100:.1f}%"],
    [regions[4], f"{region_sales[4]:,.0f}", f"{(region_sales[4]/sum(region_sales))*100:.1f}%"]
]
table = ax8.table(cellText=table_data, loc='center', cellLoc='center')
table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1, 1.5)
ax8.set_title('รายละเอียดยอดขายตามภูมิภาค')

plt.tight_layout()
plt.suptitle('Dashboard สรุปยอดขาย', fontsize=16, y=1.02)
plt.show()
````
