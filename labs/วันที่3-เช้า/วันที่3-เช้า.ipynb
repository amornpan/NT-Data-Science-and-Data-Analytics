## บทนำ

ในช่วงเช้าของวันที่ 3 เราจะเรียนรู้เกี่ยวกับหลักการสร้าง Dashboard ที่มีประสิทธิภาพ และวิธีการเลือกกราฟที่เหมาะสมสำหรับข้อมูลและความต้องการวิเคราะห์ประเภทต่างๆ ใน Power BI

## 1. ความเข้าใจพื้นฐานเกี่ยวกับ Dashboard

### 1.1 Dashboard คืออะไร?

Dashboard คือหน้าจอแสดงผลที่รวบรวมข้อมูลสำคัญและตัวชี้วัดในรูปแบบวิชวลต่างๆ เพื่อให้ผู้ใช้สามารถติดตามผลการดำเนินงาน เข้าใจแนวโน้ม และตัดสินใจบนพื้นฐานของข้อมูล

### 1.2 ประเภทของ Dashboard

1. **Dashboard เชิงปฏิบัติการ (Operational Dashboard)**
   - ติดตามการดำเนินงานประจำวัน/สัปดาห์
   - มุ่งเน้นที่ตัวชี้วัดปัจจุบัน
   - ตัวอย่าง: การติดตามยอดขายรายวัน, สถานะการจัดส่งสินค้า

2. **Dashboard เชิงกลยุทธ์ (Strategic Dashboard)**
   - ติดตามเป้าหมายระยะยาว
   - มุ่งเน้นที่ KPI และเป้าหมายขององค์กร
   - ตัวอย่าง: ผลการดำเนินงานรายไตรมาส, การเติบโตของธุรกิจ

3. **Dashboard เชิงวิเคราะห์ (Analytical Dashboard)**
   - มุ่งเน้นที่การวิเคราะห์เชิงลึก
   - สนับสนุนการตัดสินใจ
   - ตัวอย่าง: การวิเคราะห์พฤติกรรมลูกค้า, ประสิทธิภาพแคมเปญการตลาด

### 1.3 หลักการพื้นฐานในการออกแบบ Dashboard

1. **เน้นผู้ใช้เป็นศูนย์กลาง (User-Centric Approach)**
   - เข้าใจความต้องการและงานของผู้ใช้
   - ปรับ Dashboard ให้เหมาะกับระดับความรู้และความต้องการของผู้ใช้

2. **จัดลำดับข้อมูลอย่างชัดเจน (Information Hierarchy)**
   - นำเสนอข้อมูลสำคัญที่สุดก่อน
   - ใช้ขนาด, สี และตำแหน่งเพื่อแสดงความสำคัญ

3. **ความสม่ำเสมอในการออกแบบ (Visual Consistency)**
   - ใช้รูปแบบ, สี และวิชวลที่สอดคล้องกัน
   - สร้างภาษาภาพที่สม่ำเสมอ

4. **การจัดวางที่เรียบง่าย (Scannable Layout)**
   - ใช้การจัดวางตามตารางกริด
   - จัดกลุ่มข้อมูลที่เกี่ยวข้องไว้ด้วยกัน
   - ลดความรกและรักษาพื้นที่ว่าง

5. **บริบทและความเกี่ยวข้อง (Context and Relevance)**
   - ให้บริบทกับข้อมูล (เช่น การเปรียบเทียบกับช่วงเวลาก่อนหน้า)
   - แสดงเฉพาะข้อมูลที่เกี่ยวข้องกับการตัดสินใจ

## 2. อินเตอร์เฟซของ Power BI และโครงสร้างของ Dashboard

### 2.1 ส่วนประกอบหลักของ Power BI Desktop

1. **แถบริบบอน (Ribbon)** - เมนูด้านบนที่มีคำสั่งหลักในการทำงาน
2. **พื้นที่ทำงาน (Canvas)** - พื้นที่สำหรับวางวิชวลต่างๆ
3. **ตัวเลือกโหมดมุมมอง (View Selector)** - สลับระหว่างมุมมองรายงาน, ข้อมูล, และความสัมพันธ์
4. **พาเนลการสร้างวิชวล (Visualizations Pane)** - เลือกและกำหนดค่าวิชวล
5. **พาเนลข้อมูล (Fields Pane)** - เข้าถึงตารางและฟิลด์ข้อมูล

### 2.2 โครงสร้างโซนของ Dashboard

Dashboard ที่ออกแบบอย่างมีประสิทธิภาพควรมีโครงสร้างที่ชัดเจน แบ่งเป็นโซนต่างๆ ดังนี้:

1. **ส่วนหัว (ระดับศูนย์)**
   - ชื่อ Dashboard
   - ตัวกรองทั่วไป (Global Filters)
   - ปุ่มนำทาง

2. **ระดับที่หนึ่ง - KPI**
   - การ์ดแสดงเมตริกสำคัญ
   - ตัวชี้วัดหลักของธุรกิจ
   - กฎสำคัญ: เริ่มต้น Dashboard ด้วยการ์ด KPI เสมอ

3. **ระดับที่สอง - ข้อมูลสรุปและแนวโน้ม**
   - กราฟแสดงข้อมูลสรุประดับสูง
   - แนวโน้มตามเวลา
   - กราฟที่ทำหน้าที่เป็นตัวกรองสำหรับวิชวลอื่นๆ

4. **ระดับที่สาม - รายละเอียด**
   - ข้อมูลแบบละเอียด
   - ตารางและกราฟแสดงรายละเอียดเฉพาะ
   - วิชวลสนับสนุนอื่นๆ

## 3. หลักการเลือกกราฟที่เหมาะสม: เข็มทิศการสร้างวิชวล (Visualization Compass)

การเลือกกราฟที่เหมาะสมกับข้อมูลและวัตถุประสงค์ของเราเป็นสิ่งสำคัญอย่างยิ่ง เข็มทิศการสร้างวิชวล (Visualization Compass) เป็นกรอบแนวคิดที่ช่วยเราเลือกกราฟที่เหมาะสมตามวัตถุประสงค์หลักของการวิเคราะห์

### 3.1 วัตถุประสงค์หลักของการวิเคราะห์

1. **การจัดอันดับ (Ranking)** - การเปรียบเทียบค่าระหว่างหมวดหมู่
2. **พลวัต (Dynamics)** - การวิเคราะห์การเปลี่ยนแปลงตามเวลา
3. **โครงสร้าง (Structure)** - การแสดงความสัมพันธ์ระหว่างส่วนย่อยกับทั้งหมด
4. **ความสัมพันธ์ (Relationship)** - การแสดงความสัมพันธ์ระหว่างตัวแปร
5. **การกระจายตัว (Distribution)** - การแสดงการกระจายตัวของข้อมูล

### 3.2 การเลือกกราฟตามวัตถุประสงค์

#### การจัดอันดับ (Ranking)

การเปรียบเทียบค่าระหว่างหมวดหมู่ต่างๆ โดยมีเกณฑ์การเลือก:

- **กราฟแท่ง (Column Chart)**
  - เหมาะสำหรับ 3-7 หมวดหมู่
  - ดีสำหรับป้ายกำกับสั้นๆ
  - ใช้เมื่อต้องการเน้นการเปรียบเทียบค่า
  
- **กราฟแท่งแนวนอน (Bar Chart)**
  - เหมาะสำหรับ 8+ หมวดหมู่
  - ดีสำหรับป้ายกำกับยาว
  - เหมาะสำหรับการจัดอันดับจากมากไปน้อย
  
- **ตาราง (Table)**
  - เมื่อต้องการค่าที่แม่นยำ
  - เมื่อมีหลายมิติของข้อมูลที่ต้องแสดง
  - สามารถใช้การจัดรูปแบบตามเงื่อนไขเพื่อเพิ่มความชัดเจน

#### พลวัต (Dynamics)

การวิเคราะห์การเปลี่ยนแปลงตามเวลาหรือลำดับ:

- **กราฟเส้น (Line Chart)**
  - เหมาะสำหรับแนวโน้มต่อเนื่อง
  - ดีสำหรับการแสดงอัตราการเปลี่ยนแปลง
  - สามารถแสดงหลายชุดข้อมูลพร้อมกัน
  
- **กราฟพื้นที่ (Area Chart)**
  - เหมาะสำหรับแนวโน้มสะสมหรือส่วนประกอบที่เปลี่ยนแปลงตามเวลา
  - แสดงปริมาณได้ชัดเจนกว่ากราฟเส้น
  
- **กราฟแท่ง (Column Chart)**
  - เหมาะสำหรับช่วงเวลาที่ไม่ต่อเนื่อง (เช่น รายไตรมาส)
  - ดีสำหรับการเปรียบเทียบค่าในแต่ละช่วงเวลา

#### โครงสร้าง (Structure)

การแสดงความสัมพันธ์ระหว่างส่วนย่อยกับทั้งหมด:

- **กราฟวงกลม/โดนัท (Pie/Donut Chart)**
  - เหมาะสำหรับ 2-6 หมวดหมู่ที่มีความแตกต่างชัดเจน
  - ดีสำหรับแสดงสัดส่วนของทั้งหมด
  - ควรใช้เมื่อผลรวมมีความหมาย
  
- **แผนภูมิต้นไม้ (Treemap)**
  - เหมาะสำหรับหลายหมวดหมู่และมีโครงสร้างเป็นลำดับชั้น
  - แสดงทั้งขนาดและลำดับชั้นได้พร้อมกัน
  
- **กราฟแท่งซ้อน/กราฟแท่งซ้อน 100% (Stacked Bar/100% Stacked Bar)**
  - เหมาะสำหรับเปรียบเทียบส่วนประกอบภายในกลุ่มและระหว่างกลุ่ม
  - 100% Stacked เน้นสัดส่วนภายในแต่ละกลุ่ม

#### ความสัมพันธ์ (Relationship)

การวิเคราะห์ความสัมพันธ์ระหว่างตัวแปร:

- **แผนภูมิการกระจาย (Scatter Plot)**
  - แสดงความสัมพันธ์ระหว่างตัวแปร 2 ตัว
  - ดีสำหรับหาสหสัมพันธ์หรือแนวโน้ม
  
- **แผนภูมิบับเบิล (Bubble Chart)**
  - แสดงความสัมพันธ์ระหว่างตัวแปร 3 ตัว (X, Y และขนาด)
  - สามารถเพิ่มมิติที่ 4 ด้วยสี

#### ข้อมูลเชิงภูมิศาสตร์ (Geographic)

การวิเคราะห์ข้อมูลตามตำแหน่งทางภูมิศาสตร์:

- **แผนที่ (Map)**
  - แสดงข้อมูลตามตำแหน่งทางภูมิศาสตร์
  - เหมาะสำหรับการวิเคราะห์การกระจายตัวทางภูมิศาสตร์
  
- **แผนที่แบบระบายสี (Filled Map)**
  - แสดงข้อมูลรายพื้นที่ (เช่น จังหวัด, ประเทศ)
  - เหมาะสำหรับเปรียบเทียบค่าระหว่างพื้นที่

### 3.3 การเลือกกราฟอัตโนมัติใน Power BI

Power BI จะแนะนำการแสดงผลโดยอัตโนมัติตามประเภทข้อมูลที่คุณเลือกก่อน:

- หากคุณเลือกข้อมูลตัวเลขก่อน → กราฟแท่ง (Column chart)
- หากคุณเลือกข้อมูลข้อความก่อน → ตาราง (Table)
- หากคุณเลือกข้อมูลทางภูมิศาสตร์ก่อน → แผนที่ (Map)

คุณสามารถแทนที่ค่าเริ่มต้นเหล่านี้โดย:
1. สร้างคอนเทนเนอร์วิชวลว่างก่อน
2. เลือกประเภทการแสดงผลที่คุณต้องการ
3. เพิ่มฟิลด์ข้อมูลลงในพื้นที่ที่เหมาะสม

## 4. แนวปฏิบัติที่ดีและข้อผิดพลาดทั่วไปในการสร้าง Dashboard

### 4.1 แนวปฏิบัติที่ดีในการออกแบบ Dashboard

1. **เริ่มต้นจากคำถามสำคัญที่ต้องการตอบ**
   - กำหนดวัตถุประสงค์และผู้ใช้งานให้ชัดเจน
   - ระบุ KPI ที่สำคัญและต้องติดตาม

2. **ใช้หลักการออกแบบ Dashboard ที่ชัดเจน**
   - จัดวางตามหลักลำดับความสำคัญ (เริ่มจาก KPI ที่ส่วนบน)
   - สร้างตารางกริดสำหรับการจัดวาง
   - จัดกลุ่มข้อมูลที่เกี่ยวข้องไว้ด้วยกัน

3. **การใช้สีอย่างมีประสิทธิภาพ**
   - ใช้โทนสีที่สอดคล้องกับแบรนด์
   - จำกัดการใช้สีให้น้อยที่สุด (3-5 สี)
   - กำหนดสีเฉพาะสำหรับเมตริกแต่ละประเภท (เช่น ยอดขาย = สีน้ำเงิน, กำไร = สีเขียว)
   - ใช้ความเข้มของสีแสดงค่า (มากกว่า = เข้มกว่า)

4. **การจัดการตัวกรองและการโต้ตอบ**
   - วางตัวกรองในตำแหน่งที่เข้าถึงง่าย
   - ตั้งค่าการกรองแบบไขว้ที่เหมาะสม
   - สร้าง drill-down สำหรับข้อมูลเชิงลึก

5. **การเพิ่มบริบทและการตีความ**
   - แสดงการเปรียบเทียบกับเป้าหมาย/ช่วงเวลาก่อนหน้า
   - ใช้ไอคอนและสัญลักษณ์แสดงสถานะ (เช่น ขึ้น/ลง)
   - เพิ่มคำอธิบายสั้นๆ หากจำเป็น

### 4.2 ข้อผิดพลาดทั่วไปที่ควรหลีกเลี่ยง

1. **การใช้กราฟไม่เหมาะสมกับข้อมูล**
   - ใช้กราฟวงกลมสำหรับหลายหมวดหมู่ (มากกว่า 6-7 รายการ)
   - ใช้กราฟเส้นสำหรับข้อมูลที่ไม่ต่อเนื่อง
   - ใช้กราฟแท่งสำหรับรายการที่มีป้ายกำกับยาว

2. **การไม่ปรับสเกลให้เหมาะสม**
   - เริ่มแกนค่าที่ไม่ใช่ศูนย์โดยไม่มีเหตุผล (ทำให้ความแตกต่างดูมากเกินไป)
   - ใช้สเกลลอการิทึมโดยไม่ระบุให้ชัดเจน
   - ใช้สเกลที่แตกต่างกันสำหรับการเปรียบเทียบโดยตรง

3. **การใช้ภาพประกอบที่ไม่จำเป็น**
   - กราฟ 3D ที่บิดเบือนข้อมูล
   - ภาพพื้นหลังที่รบกวนการอ่านข้อมูล
   - การใช้เอฟเฟกต์การเคลื่อนไหวที่ไม่จำเป็น

4. **ความแออัดของข้อมูล**
   - ใส่ข้อมูลมากเกินไปในหนึ่งวิชวล
   - ใส่วิชวลมากเกินไปในหนึ่งหน้า
   - ขาดพื้นที่ว่างระหว่างวิชวลต่างๆ

5. **ไม่คำนึงถึงประสบการณ์ผู้ใช้**
   - ไม่มีคำอธิบายหรือชื่อที่ชัดเจน
   - ไม่มีการโต้ตอบหรือความสามารถในการกรอง
   - ไม่ออกแบบให้รองรับอุปกรณ์ที่ใช้งานจริง

## 5. การนำไปใช้งานจริง: ขั้นตอนการสร้าง Dashboard ใน Power BI

### 5.1 ขั้นตอนการสร้าง Dashboard

1. **วางแผนและออกแบบ Dashboard**
   - กำหนดวัตถุประสงค์และ KPI ที่ต้องการติดตาม
   - ระบุผู้ใช้งานและความต้องการ
   - สร้างโครงร่างของ Dashboard บนกระดาษหรือเครื่องมืออื่น

2. **นำเข้าและเตรียมข้อมูล**
   - เชื่อมต่อกับแหล่งข้อมูล
   - ทำความสะอาดและแปลงข้อมูล
   - สร้างความสัมพันธ์ระหว่างตาราง
   - สร้างคอลัมน์และตัววัดเพิ่มเติมที่จำเป็น

3. **สร้างวิชวลเริ่มต้น**
   - เริ่มต้นด้วยการสร้าง KPI cards
   - เพิ่มกราฟแสดงแนวโน้มและการเปรียบเทียบ
   - เพิ่มรายละเอียดเชิงลึกตามความจำเป็น

4. **เพิ่มตัวกรองและการโต้ตอบ**
   - เพิ่ม slicers และตัวกรองทั่วไป
   - กำหนดการกรองแบบไขว้ระหว่างวิชวล
   - สร้าง drill-down และ tooltips ที่เป็นประโยชน์

5. **ปรับแต่งและตกแต่ง**
   - จัดเรียงและปรับขนาดวิชวล
   - กำหนดโทนสีและธีมที่สอดคล้องกัน
   - เพิ่มชื่อและคำอธิบายที่ชัดเจน

6. **ทดสอบและรับฟีดแบ็ค**
   - ทดสอบกับสถานการณ์และข้อมูลจริง
   - รับฟีดแบ็คจากผู้ใช้งาน
   - ปรับปรุงตามความจำเป็น

7. **เผยแพร่และแชร์**
   - เผยแพร่ Dashboard สู่ Power BI Service
   - กำหนดสิทธิ์การเข้าถึง
   - สร้างรายงานอัตโนมัติหรือการแจ้งเตือนตามความเหมาะสม

### 5.2 แนวทางการนำเสนอ Dashboard

เมื่อต้องนำเสนอ Dashboard ให้กับผู้ใช้งานหรือผู้บริหาร ควรปฏิบัติดังนี้:

1. **เริ่มด้วยภาพรวม**
   - อธิบายวัตถุประสงค์และประโยชน์ของ Dashboard
   - นำเสนอ KPI หลักและความหมาย

2. **อธิบายวิธีการอ่านและใช้งาน**
   - อธิบายความหมายของแต่ละวิชวล
   - แสดงวิธีการใช้ตัวกรองและการโต้ตอบ

3. **นำเสนอข้อมูลเชิงลึก**
   - เน้นที่ข้อค้นพบสำคัญจากข้อมูล
   - เชื่อมโยงข้อมูลเชิงลึกกับการตัดสินใจทางธุรกิจ

4. **แสดงตัวอย่างการใช้งานจริง**
   - แสดงสถานการณ์การใช้งานจริง
   - ใช้ drill-down เพื่อตอบคำถามเฉพาะ

5. **เปิดโอกาสให้ซักถามและรับฟีดแบ็ค**
   - ตอบคำถามและข้อสงสัย
   - รับฟีดแบ็คเพื่อการปรับปรุง

## สรุปและแบบฝึกหัด

### สรุปเนื้อหา

ในช่วงเช้านี้ เราได้เรียนรู้เกี่ยวกับ:
1. พื้นฐานและประเภทของ Dashboard
2. โครงสร้างและหลักการออกแบบ Dashboard
3. การเลือกกราฟที่เหมาะสมกับข้อมูลและวัตถุประสงค์ (Visualization Compass)
4. แนวปฏิบัติที่ดีและข้อผิดพลาดที่ควรหลีกเลี่ยง
5. ขั้นตอนการสร้างและนำเสนอ Dashboard ในสถานการณ์จริง

### แบบฝึกหัด

1. **การเลือกกราฟที่เหมาะสม**
   - จากข้อมูลที่กำหนดให้ ผู้เรียนต้องเลือกกราฟที่เหมาะสมและอธิบายเหตุผล

2. **การออกแบบโครงสร้าง Dashboard**
   - ออกแบบโครงร่าง Dashboard จากโจทย์ธุรกิจที่กำหนดให้
   - ระบุ KPI, กราฟที่จะใช้ และการจัดวาง

3. **การวิเคราะห์ Dashboard**
   - วิเคราะห์ Dashboard ตัวอย่าง ระบุจุดแข็งและจุดที่ควรปรับปรุง
   - เสนอแนะการปรับปรุงเพื่อให้มีประสิทธิภาพมากขึ้น

- วิเคราะห์ข้อมูลจากสถานการณ์จริง
