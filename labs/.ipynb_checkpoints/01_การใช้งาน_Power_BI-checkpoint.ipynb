# การใช้งาน Power BI สำหรับการวิเคราะห์ข้อมูลและการสร้างการแสดงผลข้อมูล

## วัตถุประสงค์การเรียนรู้
- เข้าใจพื้นฐานและความสามารถของ Power BI
- สามารถนำเข้าและจัดการข้อมูลใน Power BI
- สามารถสร้างการแสดงผลข้อมูลและ Dashboard ที่มีประสิทธิภาพ
- เรียนรู้การแชร์และเผยแพร่รายงาน Power BI

## 1. แนะนำ Power BI

Power BI คือชุดเครื่องมือวิเคราะห์ธุรกิจของ Microsoft ที่ช่วยให้คุณสามารถเชื่อมต่อกับข้อมูลจากหลากหลายแหล่ง สร้างการแสดงผลข้อมูลที่มีประสิทธิภาพ และแชร์ข้อมูลเชิงลึกกับทั้งองค์กร

### 1.1 ส่วนประกอบของ Power BI

Power BI ประกอบด้วยองค์ประกอบหลัก 3 ส่วน:

1. **Power BI Desktop**: แอปพลิเคชันที่ติดตั้งบนเครื่องคอมพิวเตอร์ของคุณสำหรับสร้างรายงานและการแสดงผลข้อมูล
2. **Power BI Service (cloud)**: บริการออนไลน์สำหรับแชร์รายงานและ Dashboard
3. **Power BI Mobile**: แอปสำหรับอุปกรณ์พกพาเพื่อดู Dashboard และรายงานจากทุกที่

ในการอบรมนี้เราจะเน้นการใช้งาน Power BI Desktop เป็นหลัก

### 1.2 อินเตอร์เฟซและองค์ประกอบของ Power BI Desktop

เมื่อเปิด Power BI Desktop คุณจะพบกับส่วนประกอบหลักดังนี้:

1. **The ribbon (แถบเมนูด้านบน)**: ประกอบด้วยคำสั่งต่างๆ จัดกลุ่มเป็นแท็บ (Home, View, Modeling เป็นต้น)
2. **Workspace canvas (พื้นที่ทำงาน)**: พื้นที่หลักสำหรับสร้างการแสดงผลข้อมูลและรายงาน
3. **View mode selector (ตัวเลือกโหมดมุมมอง)**: อยู่ด้านซ้าย มี 3 มุมมอง - Report, Data, และ Model
4. **Visualizations pane (แผงการสร้างวิชวล)**: แสดงประเภทการแสดงผลข้อมูลที่สามารถสร้างได้และตัวควบคุมการแสดงผล
5. **Data pane (แผงข้อมูล)**: แสดงตารางและฟิลด์ข้อมูลที่นำเข้ามา

![Power BI Interface](https://www.coursereport.com/uploads/school/courses/course-report-production/16614/dynamic_power-bi-dashboard-interface-image.jpg)

## 2. การเชื่อมต่อกับแหล่งข้อมูล

Power BI สามารถเชื่อมต่อกับแหล่งข้อมูลได้หลากหลาย ตั้งแต่ไฟล์ Excel, CSV, ฐานข้อมูล SQL ไปจนถึงบริการออนไลน์ต่างๆ

### 2.1 การนำเข้าข้อมูลจากไฟล์

#### 2.1.1 นำเข้าข้อมูลจาก Excel

1. คลิกที่ "Get Data" > "Excel"
2. เลือกไฟล์ Excel ที่ต้องการนำเข้า
3. เลือกชีตหรือตารางที่ต้องการ
4. คลิก "Load" เพื่อนำเข้าข้อมูลทันที หรือ "Transform Data" เพื่อปรับแต่งข้อมูลก่อนนำเข้า

#### 2.1.2 นำเข้าข้อมูลจาก CSV

1. คลิกที่ "Get Data" > "Text/CSV"
2. เลือกไฟล์ CSV ที่ต้องการนำเข้า
3. ตรวจสอบการตั้งค่าการนำเข้า เช่น ตัวคั่น (Delimiter), การเข้ารหัส (Encoding)
4. คลิก "Load" หรือ "Transform Data"

### 2.2 การเชื่อมต่อกับฐานข้อมูล

Power BI สามารถเชื่อมต่อกับฐานข้อมูลได้หลายประเภท เช่น SQL Server, MySQL, PostgreSQL

#### 2.2.1 เชื่อมต่อกับ SQL Server

1. คลิกที่ "Get Data" > "Database" > "SQL Server"
2. ป้อนชื่อเซิร์ฟเวอร์และฐานข้อมูล
3. เลือกโหมดการเชื่อมต่อ: Import หรือ DirectQuery
4. เลือกตารางที่ต้องการนำเข้า
5. คลิก "Load" หรือ "Transform Data"

### 2.3 การเชื่อมต่อกับบริการออนไลน์

Power BI สามารถเชื่อมต่อกับบริการออนไลน์เช่น SharePoint, Google Analytics และอื่นๆ

ตัวอย่างการเชื่อมต่อกับ SharePoint:
1. คลิกที่ "Get Data" > "SharePoint Online List"
2. ป้อน URL ของไซต์ SharePoint
3. เลือกรายการที่ต้องการนำเข้า
4. คลิก "Load" หรือ "Transform Data"

### 2.4 การรวมข้อมูลจากหลายแหล่ง

Power BI สามารถรวมข้อมูลจากหลายแหล่งเข้าด้วยกัน โดยสามารถใช้ Power Query Editor หรือสร้างความสัมพันธ์ระหว่างตารางในโมเดลข้อมูล

## 3. การจัดการและปรับแต่งข้อมูล

### 3.1 การใช้งาน Power Query Editor

Power Query Editor เป็นเครื่องมือที่ทรงพลังสำหรับการปรับแต่งข้อมูลก่อนนำมาวิเคราะห์และแสดงผล

วิธีเปิด Power Query Editor:
1. คลิกที่ "Transform Data" ในแถบเมนู หรือ
2. เลือก "Transform Data" ในขั้นตอนการนำเข้าข้อมูล

![Power Query Editor](https://www.c-sharpcorner.com/article/getting-started-with-the-power-bi-query-editor2/Images/1.png)

### 3.2 การทำความสะอาดข้อมูล

ใน Power Query Editor คุณสามารถทำความสะอาดข้อมูลได้ดังนี้:

1. **ลบข้อมูลที่ขาดหายไป (Missing values)**:
   - คลิกขวาที่คอลัมน์ > "Remove Empty"
   - หรือใช้ตัวกรอง (Filter) เพื่อลบแถวที่มีค่า null

2. **แทนที่ค่า**:
   - คลิกขวาที่คอลัมน์ > "Replace Values"
   - ป้อนค่าที่ต้องการค้นหาและค่าที่ต้องการแทนที่

3. **แก้ไขประเภทข้อมูล**:
   - คลิกไอคอนประเภทข้อมูลที่ส่วนหัวของคอลัมน์
   - เลือกประเภทข้อมูลที่ถูกต้อง เช่น Text, Number, Date

### 3.3 การปรับโครงสร้างข้อมูล

Power Query Editor มีเครื่องมือสำหรับปรับโครงสร้างข้อมูล:

1. **Pivot/Unpivot** เพื่อเปลี่ยนข้อมูลจากแนวนอนเป็นแนวตั้งหรือกลับกัน
2. **Split Column** เพื่อแยกข้อมูลในคอลัมน์เดียวออกเป็นหลายคอลัมน์
3. **Merge Columns** เพื่อรวมข้อมูลจากหลายคอลัมน์เข้าด้วยกัน
4. **Transpose** เพื่อสลับแถวกับคอลัมน์

### 3.4 การสร้างคอลัมน์เพิ่มเติม

คุณสามารถสร้างคอลัมน์ใหม่ใน Power Query Editor:

1. **Add Custom Column**: 
   - คลิกที่ "Add Column" > "Custom Column"
   - เขียนสูตรโดยใช้ภาษา M
   
2. **Conditional Column**:
   - คลิกที่ "Add Column" > "Conditional Column"
   - กำหนดเงื่อนไขและค่าที่จะใส่เมื่อเป็นจริงหรือเท็จ

## 4. การสร้างโมเดลข้อมูล

### 4.1 ความสัมพันธ์ระหว่างตาราง

หากมีข้อมูลหลายตาราง คุณสามารถสร้างความสัมพันธ์ระหว่างตารางได้:

1. ไปที่มุมมอง Model (โหมด Model)
2. ลากฟิลด์จากตารางหนึ่งไปยังอีกตารางหนึ่งเพื่อสร้างความสัมพันธ์
3. หรือคลิกที่ "Manage Relationships" ในแถบเมนู Home เพื่อสร้างและจัดการความสัมพันธ์

![Data Relationships](https://learn.microsoft.com/en-us/training/wwl-data-ai/analyze-data-power-bi/media/relationship-view-8ca064f6.png)

### 4.2 การสร้างตาบเกลา (Hierarchies)

ตาบเกลาช่วยให้คุณสามารถเจาะลึกข้อมูลได้หลายระดับ:

1. ในแผง Data คลิกขวาที่ฟิลด์ที่ต้องการเป็นระดับบนสุด
2. เลือก "Create hierarchy"
3. คลิกขวาที่ฟิลด์อื่นๆ และเลือก "Add to hierarchy" เพื่อเพิ่มระดับถัดไป

ตัวอย่าง:
- สร้างตาบเกลาของเวลา: ปี > ไตรมาส > เดือน > วัน
- สร้างตาบเกลาของสถานที่: ภูมิภาค > ประเทศ > จังหวัด > เมือง

### 4.3 การสร้างเมทริกซ์ (Measures) ด้วย DAX

DAX (Data Analysis Expressions) คือภาษาสูตรที่ใช้ใน Power BI สำหรับสร้างการคำนวณขั้นสูง:

1. คลิกขวาที่ตารางในแผง Data และเลือก "New measure"
2. เขียนสูตร DAX
3. กด Enter เพื่อสร้างเมทริกซ์

ตัวอย่างสูตร DAX:
```
Total Sales = SUM(Sales[Amount])
Average Sales = AVERAGE(Sales[Amount])
YoY Growth % = (SUM(Sales[Amount]) - CALCULATE(SUM(Sales[Amount]), SAMEPERIODLASTYEAR(Calendar[Date]))) / CALCULATE(SUM(Sales[Amount]), SAMEPERIODLASTYEAR(Calendar[Date]))
```

### 4.4 การคำนวณคอลัมน์ (Calculated Columns)

คุณสามารถสร้างคอลัมน์ใหม่จากการคำนวณในโมเดลข้อมูลได้:

1. ไปที่มุมมอง Data หรือ Model
2. คลิกขวาที่ตารางและเลือก "New column"
3. เขียนสูตร DAX
4. กด Enter เพื่อสร้างคอลัมน์

## 5. การสร้างการแสดงผลข้อมูล

### 5.1 การสร้างแผนภูมิพื้นฐาน

Power BI มีแผนภูมิและการแสดงผลข้อมูลหลายรูปแบบให้เลือกใช้:

#### 5.1.1 แผนภูมิแท่ง (Column Chart)

เหมาะสำหรับเปรียบเทียบค่าระหว่างหมวดหมู่ต่างๆ:

1. คลิกที่ไอคอนแผนภูมิแท่งในแผง Visualizations
2. ลากฟิลด์หมวดหมู่ไปยัง "Axis" (เช่น ชื่อผลิตภัณฑ์, เดือน)
3. ลากฟิลด์ตัวเลขไปยัง "Values" (เช่น ยอดขาย, จำนวน)

![Column Chart](https://learn.microsoft.com/th-th/power-bi/visuals/media/power-bi-visualization-types-for-reports-and-q-and-a/clustered-column-chart.png)

#### 5.1.2 แผนภูมิเส้น (Line Chart)

เหมาะสำหรับแสดงแนวโน้มตามเวลา:

1. คลิกที่ไอคอนแผนภูมิเส้นในแผง Visualizations
2. ลากฟิลด์เวลาไปยัง "Axis" (เช่น วันที่, เดือน, ปี)
3. ลากฟิลด์ตัวเลขไปยัง "Values" (เช่น ยอดขาย)

![Line Chart](https://learn.microsoft.com/th-th/power-bi/visuals/media/power-bi-visualization-types-for-reports-and-q-and-a/line-chart.png)

#### 5.1.3 แผนภูมิวงกลม (Pie Chart)

เหมาะสำหรับแสดงสัดส่วนของทั้งหมด:

1. คลิกที่ไอคอนแผนภูมิวงกลมในแผง Visualizations
2. ลากฟิลด์หมวดหมู่ไปยัง "Legend" (เช่น ประเภทสินค้า)
3. ลากฟิลด์ตัวเลขไปยัง "Values" (เช่น ยอดขาย)

**หมายเหตุ**: ควรใช้แผนภูมิวงกลมเมื่อมีหมวดหมู่ไม่เกิน 5-7 รายการ เพื่อให้อ่านง่าย

![Pie Chart](https://learn.microsoft.com/th-th/power-bi/visuals/media/power-bi-visualization-types-for-reports-and-q-and-a/pie-chart.png)

### 5.2 การสร้างตาราง (Tables) และเมทริกซ์ (Matrix)

#### 5.2.1 ตาราง (Table)

ตารางแสดงข้อมูลในรูปแบบที่คล้ายกับ Excel:

1. คลิกที่ไอคอนตารางในแผง Visualizations
2. ลากฟิลด์ที่ต้องการแสดงไปยัง "Values"

![Table](https://learn.microsoft.com/th-th/power-bi/visuals/media/power-bi-visualization-types-for-reports-and-q-and-a/table-visual.png)

#### 5.2.2 เมทริกซ์ (Matrix)

เมทริกซ์คล้ายกับ Pivot Table ใน Excel:

1. คลิกที่ไอคอนเมทริกซ์ในแผง Visualizations
2. ลากฟิลด์แถวไปยัง "Rows" (เช่น ประเทศ, เมือง)
3. ลากฟิลด์คอลัมน์ไปยัง "Columns" (เช่น ปี, ไตรมาส)
4. ลากฟิลด์ตัวเลขไปยัง "Values" (เช่น ยอดขาย)

### 5.3 การใช้งาน Map และการแสดงผลข้อมูลเชิงพื้นที่

Power BI มีการแสดงผลข้อมูลเชิงพื้นที่หลายรูปแบบ:

#### 5.3.1 Map (แผนที่)

1. คลิกที่ไอคอนแผนที่ในแผง Visualizations
2. ลากฟิลด์ตำแหน่ง (เช่น ประเทศ, เมือง, รหัสไปรษณีย์) ไปยัง "Location"
3. ลากฟิลด์ตัวเลขไปยัง "Size" (เช่น ยอดขาย)

![Map](https://learn.microsoft.com/th-th/power-bi/visuals/media/power-bi-visualization-types-for-reports-and-q-and-a/map.png)

#### 5.3.2 Filled Map (แผนที่แบบระบายสี)

1. คลิกที่ไอคอน Filled Map ในแผง Visualizations
2. ลากฟิลด์ตำแหน่งไปยัง "Location"
3. ลากฟิลด์ตัวเลขไปยัง "Color saturation"

### 5.4 การปรับแต่งการแสดงผลข้อมูล

คุณสามารถปรับแต่งการแสดงผลข้อมูลได้โดย:

1. เลือกการแสดงผลข้อมูลที่ต้องการปรับแต่ง
2. ใช้แผง Format (ไอคอนลูกกลิ้งสี) ในแผง Visualizations
3. ปรับแต่งสี ขนาด ฟอนต์ และอื่นๆ

## 6. การสร้างแผงควบคุม (Dashboard) และรายงาน (Report)

### 6.1 การจัดวางการแสดงผลข้อมูล

การจัดวางที่ดีช่วยให้ผู้ใช้เข้าใจข้อมูลได้ง่าย:

1. คลิกและลากการแสดงผลข้อมูลเพื่อจัดตำแหน่ง
2. ปรับขนาดโดยลากที่มุมหรือขอบ
3. จัดเรียงแบบกริดโดยคลิกขวา > "Align"
4. จัดกลุ่มการแสดงผลข้อมูลที่เกี่ยวข้องเข้าด้วยกัน

![Dashboard Example](https://learn.microsoft.com/en-us/power-bi/consumer/mobile/media/mobile-windows-10-app-dashboard-optimized/power-bi-phone-dashboard-optimized.png)

### 6.2 การสร้างและใช้งาน Slicers และ Filters

#### 6.2.1 Slicers

Slicers ช่วยให้ผู้ใช้สามารถกรองข้อมูลได้อย่างง่ายดาย:

1. คลิกที่ไอคอน Slicer ในแผง Visualizations
2. ลากฟิลด์ที่ต้องการกรองไปยัง "Field"
3. ปรับแต่งรูปแบบของ Slicer ตามต้องการ

![Slicer Example](https://learn.microsoft.com/th-th/power-bi/visuals/media/power-bi-visualization-slicers/power-bi-slicer-date.png)

#### 6.2.2 Filters

Filters ให้ความยืดหยุ่นในการกรองข้อมูล:

1. ใช้แผง Filters ทางด้านขวา
2. สามารถตั้งค่า Filters ได้ 3 ระดับ:
   - Visual level: กรองเฉพาะการแสดงผลข้อมูลที่เลือก
   - Page level: กรองทั้งหน้า
   - Report level: กรองทั้งรายงาน

### 6.3 การสร้างการโต้ตอบระหว่างการแสดงผลข้อมูล (Interactions)

คุณสามารถกำหนดว่าการเลือกข้อมูลในการแสดงผลหนึ่งจะส่งผลต่อการแสดงผลอื่นอย่างไร:

1. เลือกการแสดงผลข้อมูลที่ต้องการตั้งค่า
2. ไปที่แท็บ Format > Edit interactions
3. คลิกที่ไอคอน Filter, Highlight หรือ None บนการแสดงผลอื่นๆ เพื่อกำหนดปฏิสัมพันธ์

### 6.4 การสร้าง Bookmarks และ Drillthrough

#### 6.4.1 Bookmarks

Bookmarks ช่วยให้คุณบันทึกสถานะของรายงานและสลับระหว่างมุมมองต่างๆ:

1. ไปที่ View > Bookmarks pane
2. จัดการแสดงผลตามต้องการ
3. คลิก "Add" ในแผง Bookmarks เพื่อบันทึกสถานะ

#### 6.4.2 Drillthrough

Drillthrough ช่วยให้ผู้ใช้สามารถดูรายละเอียดของข้อมูลได้:

1. สร้างหน้ารายงานใหม่สำหรับรายละเอียด
2. ไปที่แผง Fields > เลือกฟิลด์ที่ต้องการใช้เป็น Drillthrough field
3. คลิกขวาและเลือก "Add to Drillthrough"

## 7. การแชร์และเผยแพร่รายงาน Power BI

### 7.1 การเผยแพร่ไปยัง Power BI Service

เมื่อสร้างรายงานเสร็จแล้ว คุณสามารถเผยแพร่ไปยัง Power BI Service:

1. คลิกที่ "Publish" ในแถบเมนู
2. เข้าสู่ระบบด้วยบัญชี Power BI (ใช้บัญชี Microsoft)
3. เลือก Workspace ที่ต้องการเผยแพร่

### 7.2 การแชร์รายงานและ Dashboard

เมื่อรายงานอยู่ใน Power BI Service แล้ว คุณสามารถแชร์กับผู้อื่นได้:

1. เปิดรายงานหรือ Dashboard ใน Power BI Service
2. คลิกที่ "Share"
3. ป้อนอีเมลของผู้ที่ต้องการแชร์
4. ตั้งค่าสิทธิ์การเข้าถึง
5. คลิก "Share"

### 7.3 การฝังรายงานในเว็บไซต์

คุณสามารถฝังรายงาน Power BI ในเว็บไซต์หรือแอปพลิเคชันได้:

1. เปิดรายงานใน Power BI Service
2. คลิกที่ "File" > "Embed"
3. เลือกประเภทการฝัง (Public / Secure)
4. คัดลอกโค้ด HTML ไปใช้ในเว็บไซต์

## 8. แบบฝึกหัด: การสร้างรายงานวิเคราะห์ข้อมูลขาย

### 8.1 นำเข้าข้อมูลขายตัวอย่าง

1. ดาวน์โหลดไฟล์ข้อมูลขายตัวอย่าง (Sales_Data.xlsx)
2. เปิด Power BI Desktop
3. คลิกที่ "Get Data" > "Excel"
4. เลือกไฟล์ Sales_Data.xlsx
5. เลือกชีตที่ต้องการและคลิก "Load"

### 8.2 สร้างการแสดงผลข้อมูลที่แสดงแนวโน้มของยอดขาย

1. สร้างแผนภูมิเส้นแสดงยอดขายตามเวลา
2. สร้างแผนภูมิแท่งแสดงยอดขายตามหมวดหมู่สินค้า
3. สร้างเกจ (Gauge) แสดงยอดขายเทียบกับเป้าหมาย

### 8.3 วิเคราะห์ผลิตภัณฑ์ที่ขายดีที่สุดและภูมิภาคที่มียอดขายสูงสุด

1. สร้างแผนภูมิแท่งแนวนอนแสดงผลิตภัณฑ์ที่ขายดีที่สุด
2. สร้างแผนที่แสดงยอดขายตามภูมิภาค
3. สร้างตารางแสดงรายละเอียดของผลิตภัณฑ์ที่ขายดีที่สุด

### 8.4 สร้างแผงควบคุมที่สามารถกรองข้อมูลตามช่วงเวลาและผลิตภัณฑ์

1. เพิ่ม Slicer สำหรับกรองตามช่วงเวลา (ปี, ไตรมาส, เดือน)
2. เพิ่ม Slicer สำหรับกรองตามผลิตภัณฑ์หรือหมวดหมู่ผลิตภัณฑ์
3. ตั้งค่า Interactions ระหว่างการแสดงผลข้อมูลต่างๆ
4. ตกแต่งแผงควบคุมให้สวยงามและใช้งานง่าย

## สรุป

Power BI เป็นเครื่องมือที่ทรงพลังสำหรับการวิเคราะห์ข้อมูลและสร้างการแสดงผลข้อมูล ซึ่งช่วยให้องค์กรสามารถนำข้อมูลมาสร้างคุณค่าทางธุรกิจได้อย่างมีประสิทธิภาพ

ในบทเรียนนี้เราได้เรียนรู้:
- พื้นฐานและส่วนประกอบของ Power BI
- การนำเข้าและจัดการข้อมูล
- การสร้างโมเดลข้อมูล
- การสร้างการแสดงผลข้อมูลและ Dashboard
- การแชร์และเผยแพร่รายงาน

ทักษะเหล่านี้จะช่วยให้คุณสามารถสร้างรายงานและ Dashboard ที่มีประสิทธิภาพเพื่อสนับสนุนการตัดสินใจในองค์กรของคุณ
