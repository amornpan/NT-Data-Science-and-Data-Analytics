## 4. หลักการเลือกใช้ Visualization ที่เหมาะสม

### 4.1 หลักการ RAG (Red-Amber-Green) ในการแสดงผลข้อมูล

**หลักการ RAG** เป็นแนวทางในการใช้สีเพื่อสื่อสารสถานะของข้อมูล:
- **สีแดง (Red)**: แสดงถึงสิ่งที่ต่ำกว่าเป้าหมาย หรือมีความเสี่ยง
- **สีเหลือง (Amber)**: แสดงถึงสิ่งที่ยังไม่บรรลุเป้าหมายแต่อยู่ในช่วงที่ยอมรับได้
- **สีเขียว (Green)**: แสดงถึงสิ่งที่บรรลุเป้าหมาย หรือสถานะที่ดี

**การประยุกต์ใช้ในการวิเคราะห์ข้อมูลการจ้างงาน**:
- ใช้สีแดงแสดงการลดลงของจำนวนพนักงาน หรืออัตราการเปลี่ยนแปลงเป็นลบ
- ใช้สีเหลืองแสดงการเปลี่ยนแปลงเล็กน้อย หรืออยู่ในช่วง ±1-2%
- ใช้สีเขียวแสดงการเพิ่มขึ้นของจำนวนพนักงาน หรืออัตราการเปลี่ยนแปลงเป็นบวก

### 4.2 การเลือกใช้ Visualization ตามประเภทของข้อมูล

#### 4.2.1 การเปรียบเทียบค่าระหว่างหมวดหมู่
- **Bar/Column Chart**: เปรียบเทียบค่าระหว่างภูมิภาค จังหวัด หรือเพศ
- **Clustered/Stacked Bar Chart**: เปรียบเทียบค่าย่อยภายในแต่ละหมวดหมู่

**เหมาะกับข้อมูล**: จำนวนพนักงานแยกตามภูมิภาค, จำนวนพนักงานแยกตามเพศในแต่ละภูมิภาค

#### 4.2.2 การแสดงสัดส่วน
- **Pie/Donut Chart**: แสดงสัดส่วนของแต่ละหมวดหมู่ (ควรใช้เมื่อมีหมวดหมู่ไม่เกิน 5-7 หมวดหมู่)
- **Treemap**: แสดงสัดส่วนเชิงลำดับชั้น เช่น ภูมิภาค > จังหวัด

**เหมาะกับข้อมูล**: สัดส่วนพนักงานตามเพศ, สัดส่วนพนักงานตามภูมิภาค

#### 4.2.3 การแสดงแนวโน้มตามช่วงเวลา
- **Line Chart**: แสดงแนวโน้มต่อเนื่องตามช่วงเวลา
- **Area Chart**: แสดงแนวโน้มและสัดส่วนตามช่วงเวลา
- **Column Chart**: แสดงการเปรียบเทียบค่าตามช่วงเวลาที่ไม่ต่อเนื่อง

**เหมาะกับข้อมูล**: แนวโน้มการจ้างงานรายไตรมาส, อัตราการเปลี่ยนแปลง QoQ/YoY

#### 4.2.4 การแสดงข้อมูลเชิงภูมิศาสตร์
- **Map**: แสดงการกระจายตัวของข้อมูลตามพื้นที่
- **Filled Map**: แสดงความเข้มของข้อมูลตามพื้นที่ด้วยสี

**เหมาะกับข้อมูล**: จำนวนพนักงานตามจังหวัด, การเปลี่ยนแปลงของจำนวนพนักงานตามภูมิภาค

#### 4.2.5 การแสดงความสัมพันธ์
- **Scatter Chart**: แสดงความสัมพันธ์ระหว่างตัวแปรสองตัว
- **Bubble Chart**: แสดงความสัมพันธ์ระหว่างตัวแปรสามตัว (X, Y, ขนาด)

**เหมาะกับข้อมูล**: ความสัมพันธ์ระหว่างจำนวนพนักงานชายและหญิง, ความสัมพันธ์ระหว่างจำนวนพนักงานและอัตราการเปลี่ยนแปลง

#### 4.2.6 การแสดงการกระจายตัว
- **Box and Whisker Plot**: แสดงการกระจายตัวของข้อมูล
- **Histogram**: แสดงความถี่ของข้อมูลตามช่วง

**เหมาะกับข้อมูล**: การกระจายตัวของจำนวนพนักงานในแต่ละจังหวัด, การกระจายตัวของอัตราการเปลี่ยนแปลง

### 4.3 การออกแบบ Dashboard ที่มีประสิทธิภาพ

1. **ใช้หลัก Design Hierarchy**:
   - จัดวางข้อมูลสำคัญไว้ด้านบนซ้าย (ตำแหน่งที่ผู้ใช้มักจะอ่านก่อน)
   - ใช้ขนาดเพื่อแสดงความสำคัญ (ส่วนที่สำคัญควรมีขนาดใหญ่กว่า)
   - ใช้สีเพื่อดึงดูดความสนใจไปยังข้อมูลสำคัญ

2. **จัดกลุ่มข้อมูลที่เกี่ยวข้องกัน**:
   - ใช้กรอบหรือพื้นหลังเพื่อแบ่งกลุ่มข้อมูล
   - จัดวางกราฟที่เกี่ยวข้องกันไว้ใกล้กัน

3. **ใช้ Interactions ระหว่าง Visualizations**:
   - Cross-filtering: การคลิกที่กราฟหนึ่งจะกรองข้อมูลในกราฟอื่น
   - Drill-through: การคลิกเพื่อดูรายละเอียดเพิ่มเติม
   - Tooltips: แสดงข้อมูลเพิ่มเติมเมื่อวางเมาส์เหนือกราฟ

4. **ใช้ Consistent Color Palette**:
   - กำหนดชุดสีที่สอดคล้องกันทั้ง Dashboard
   - ใช้สีเดียวกันสำหรับหมวดหมู่เดียวกันในทุกกราฟ
   - ใช้สีตามหลัก RAG เมื่อต้องการแสดงสถานะ

## 5. การสร้าง Measures และ Calculated Columns ด้วย DAX

### 5.1 Measures ที่ควรสร้าง

#### 5.1.1 Total Employees
```dax
Total Employees = SUM(employee_stats[VALUE])
```

#### 5.1.2 Male Employees
```dax
Male Employees = CALCULATE(SUM(employee_stats[VALUE]), employee_stats[SEX] = "ชาย")
```

#### 5.1.3 Female Employees
```dax
Female Employees = CALCULATE(SUM(employee_stats[VALUE]), employee_stats[SEX] = "หญิง")
```

#### 5.1.4 Gender Ratio
```dax
Male to Female Ratio = DIVIDE([Male Employees], [Female Employees], 0)
```

#### 5.1.5 YoY Change
```dax
YoY Change = 
VAR CurrentValue = [Total Employees]
VAR PriorValue = CALCULATE(
    [Total Employees],
    DATEADD('DateTable'[Date], -1, YEAR)
)
RETURN
    DIVIDE(CurrentValue - PriorValue, PriorValue, 0)
```

#### 5.1.6 QoQ Change
```dax
QoQ Change = 
VAR CurrentValue = [Total Employees]
VAR PriorValue = CALCULATE(
    [Total Employees],
    DATEADD('DateTable'[Date], -3, MONTH)
)
RETURN
    DIVIDE(CurrentValue - PriorValue, PriorValue, 0)
```

#### 5.1.7 Rolling Average 4Q
```dax
Rolling Average 4Q = 
AVERAGEX(
    DATESINPERIOD(
        'DateTable'[Date],
        MAX('DateTable'[Date]),
        -12,
        MONTH
    ),
    [Total Employees]
)
```

### 5.2 Calculated Columns ที่ควรสร้าง

#### 5.2.1 Year-Quarter (ถ้ายังไม่มี)
```dax
Year-Quarter = employee_stats[YEAR] & "-Q" & employee_stats[QUARTER_NUM]
```

#### 5.2.2 PercentOfRegion (ถ้ายังไม่มี)
```dax
PercentOfRegion = 
DIVIDE(
    employee_stats[VALUE],
    CALCULATE(
        SUM(employee_stats[VALUE]),
        ALLEXCEPT(employee_stats, employee_stats[YEAR], employee_stats[QUARTER], employee_stats[REGION])
    ),
    0
)
```

### 5.3 การสร้าง Calculated Tables

## 6. เทคนิคการสร้าง Dashboard ที่มีประสิทธิภาพ

### 6.1 การเพิ่มประสิทธิภาพ (Performance Optimization)

1. **ลดขนาดข้อมูล**:
   - นำเข้าเฉพาะคอลัมน์ที่จำเป็น
   - ใช้ Filter ตั้งแต่ขั้นตอนการนำเข้าข้อมูล
   - ลดความละเอียดของข้อมูล (ถ้าไม่จำเป็น)

2. **ปรับปรุงการสร้าง Relationships**:
   - ใช้ Star Schema แทน Snowflake Schema
   - หลีกเลี่ยงความสัมพันธ์แบบ Many-to-Many
   - ใช้ Bi-directional filtering เฉพาะเมื่อจำเป็น

3. **ปรับปรุง DAX Measures**:
   - ใช้ Variables เพื่อเก็บค่าที่ต้องคำนวณซ้ำ
   - ใช้ฟังก์ชัน CALCULATE ร่วมกับฟิลเตอร์ที่เหมาะสม
   - หลีกเลี่ยงการคำนวณที่ซับซ้อนเกินความจำเป็น

### 6.2 การสร้าง Interactive Dashboard

1. **Slicers และ Filters**:
   - ใช้ Slicers สำหรับการกรองที่ใช้บ่อย เช่น ปี ไตรมาส ภูมิภาค
   - ใช้ Filter Pane สำหรับการกรองที่ใช้น้อยกว่า
   - ใช้ Sync Slicers ระหว่างหน้าต่างๆ

2. **Drill-through และ Drill-down**:
   - ตั้งค่า Drill-through เพื่อให้ผู้ใช้สามารถคลิกเพื่อดูรายละเอียดเพิ่มเติม
   - ใช้ Hierarchy ใน Dimensions เพื่อให้ผู้ใช้สามารถ Drill-down ได้ เช่น ภูมิภาค > จังหวัด

3. **Tooltips และ Fields**:
   - ใช้ Tooltips เพื่อแสดงข้อมูลเพิ่มเติมเมื่อวางเมาส์เหนือกราฟ
   - สร้าง Tooltip Pages เพื่อแสดงกราฟเพิ่มเติมในรูปแบบ Tooltip

4. **Bookmarks และ Buttons**:
   - ใช้ Bookmarks เพื่อบันทึกมุมมองต่างๆ ของ Dashboard
   - ใช้ Buttons เพื่อให้ผู้ใช้สามารถสลับระหว่างมุมมองต่างๆ ได้

### 6.3 การนำเสนอข้อมูลอย่างมีประสิทธิภาพ

1. **การใช้สี**:
   - ใช้สีที่มีความหมาย (ตามหลัก RAG)
   - ใช้สีที่คำนึงถึงผู้ใช้ที่มีปัญหาในการมองเห็นสี (Color Blindness)
   - ใช้สีที่สอดคล้องกันทั้ง Dashboard

2. **การใช้ Typography**:
   - ใช้ฟอนต์ที่อ่านง่าย
   - ใช้ขนาดฟอนต์ที่เหมาะสม (ไม่เล็กหรือใหญ่เกินไป)
   - ใช้ความเข้มของฟอนต์เพื่อแสดงลำดับความสำคัญ

3. **การจัดวาง Layout**:
   - ใช้ Grid Layout เพื่อให้ Dashboard ดูเป็นระเบียบ
   - จัดวางองค์ประกอบตามลำดับความสำคัญ (Z-Pattern หรือ F-Pattern)
   - ใช้พื้นที่ว่างอย่างเหมาะสม (White Space)

4. **การใช้ Titles และ Labels**:
   - ใช้ชื่อที่ชัดเจนและเข้าใจง่าย
   - ใช้คำอธิบายเพิ่มเติมเมื่อจำเป็น
   - ใช้ไอคอนหรือสัญลักษณ์ที่เป็นสากล

### 6.4 การออกแบบสำหรับผู้ใช้ต่างระดับ

1. **ผู้บริหารระดับสูง (Executive)**:

   - ใช้สีตามหลัก RAG เพื่อแสดงสถานะ

2. **ผู้บริหารระดับกลาง (Manager)**:

   - เพิ่มความสามารถในการ Drill-down
   - มี Filters ที่หลากหลายขึ้น
   - แสดงการเปรียบเทียบกับเป้าหมาย

3. **นักวิเคราะห์ (Analyst)**:
   - แสดงข้อมูลเชิงลึก
   - เพิ่มความสามารถในการสำรวจข้อมูล (Exploration)
   - มีกราฟที่ซับซ้อนขึ้น เช่น Scatter Plot, Heatmap
   - สามารถดูข้อมูลดิบหรือรายละเอียดเพิ่มเติมได้

## ภาคผนวก: ตัวอย่าง DAX สำหรับการวิเคราะห์เพิ่มเติม

### A.1 การวิเคราะห์แนวโน้ม

#### A.1.1 CAGR (Compound Annual Growth Rate)
```dax
CAGR = 
VAR StartValue = CALCULATE([Total Employees], FIRSTDATE('DateTable'[Date]))
VAR EndValue = CALCULATE([Total Employees], LASTDATE('DateTable'[Date]))
VAR YearCount = DATEDIFF(FIRSTDATE('DateTable'[Date]), LASTDATE('DateTable'[Date]), YEAR)
RETURN
    POWER(DIVIDE(EndValue, StartValue), 1/YearCount) - 1
```

#### A.1.2 Moving Average
```dax
Moving Average 2Y = 
AVERAGEX(
    DATESINPERIOD(
        'DateTable'[Date],
        MAX('DateTable'[Date]),
        -24,
        MONTH
    ),
    [Total Employees]
)
```

### A.2 การวิเคราะห์เชิงเปรียบเทียบ

#### A.2.1 Region Vs National Average
```dax
Region vs National Avg = 
DIVIDE(
    [Total Employees],
    CALCULATE(
        [Total Employees],
        ALL(employee_stats[REGION])
    ) / DISTINCTCOUNT(employee_stats[REGION])
)
```

#### A.2.2 Gender Balance Index
```dax
Gender Balance Index = 
1 - ABS(0.5 - DIVIDE([Male Employees], [Total Employees])) * 2
```

### A.3 ตัวชี้วัดประสิทธิภาพ (KPIs)

#### A.3.1 YoY Growth Target
```dax
YoY Growth Target Status = 
IF([YoY Change] >= 0.05, "Green",
   IF([YoY Change] >= 0.02, "Amber", "Red")
)
```

#### A.3.2 Gender Diversity Target
```dax
Gender Diversity Status = 
IF([Gender Balance Index] >= 0.9, "Green",
   IF([Gender Balance Index] >= 0.8, "Amber", "Red")
)
```

### A.4 การวิเคราะห์เชิงสถิติ

#### A.4.1 Standard Deviation
```dax
StdDev Employees = 
STDEVX.P(employee_stats, employee_stats[VALUE])
```

#### A.4.2 Z-Score (Standardized Value)
```dax
Z-Score = 
DIVIDE(
    employee_stats[VALUE] - AVERAGE(employee_stats[VALUE]),
    [StdDev Employees]
)
```

#### A.4.3 Percentile Rank
```dax
Percentile Rank = 
PERCENTILEX.INC(
    ALL(employee_stats),
    employee_stats[VALUE],
    employee_stats[VALUE]
)
```

การนำ DAX Expressions เหล่านี้ไปใช้จะช่วยเพิ่มความสามารถในการวิเคราะห์ข้อมูลให้ลึกซึ้งยิ่งขึ้น สามารถค้นหาข้อมูลเชิงลึกที่ซ่อนอยู่ในข้อมูลการจ้างงานได้
