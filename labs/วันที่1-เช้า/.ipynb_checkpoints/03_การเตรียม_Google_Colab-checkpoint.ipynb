# การเตรียมความพร้อมสำหรับ Google Colab

## วัตถุประสงค์การเรียนรู้
- เข้าใจว่า Google Colab คืออะไรและทำไมถึงใช้งานได้ดีกับ Data Science
- สามารถเริ่มต้นใช้งาน Google Colab ได้
- เรียนรู้วิธีการจัดการไฟล์และการโหลดข้อมูลใน Google Colab
- เข้าใจการทำงานกับ Notebook และเซลล์ประเภทต่างๆ

## Google Colab คืออะไร?

Google Colab (หรือเรียกเต็มๆ ว่า Google Colaboratory) เป็นบริการของ Google ที่อนุญาตให้ผู้ใช้เขียนและรันโค้ด Python บนเว็บเบราว์เซอร์โดยไม่ต้องติดตั้งโปรแกรมใดๆ

**ประโยชน์ของ Google Colab:**
- **ใช้งานฟรี**: ไม่มีค่าใช้จ่ายในการเริ่มต้นใช้งาน
- **ไม่ต้องติดตั้ง**: ทำงานบนคลาวด์ เข้าถึงได้จากทุกที่
- **แบ่งปันง่าย**: แชร์ไฟล์ให้ผู้อื่นได้ง่ายผ่านลิงก์
- **ไลบรารีพร้อมใช้**: มีไลบรารียอดนิยมติดตั้งไว้ให้แล้ว เช่น NumPy, Pandas, Matplotlib
- **ทรัพยากรฟรี**: ให้ใช้ CPU, GPU และ RAM ฟรี (มีข้อจำกัด)
- **จัดเก็บใน Google Drive**: บันทึกงานไว้ใน Google Drive ของคุณ

## การเริ่มต้นใช้งาน Google Colab

### 1. การเข้าถึง Google Colab
1. เข้าสู่เว็บไซต์ [colab.research.google.com](https://colab.research.google.com)
2. ลงชื่อเข้าใช้ด้วยบัญชี Google ของคุณ
3. คลิก "New Notebook" เพื่อสร้างโน้ตบุ๊กใหม่

### 2. ส่วนประกอบของหน้าจอ
- **เมนูหลัก**: ด้านบนของหน้าจอมีเมนูต่างๆ เช่น File, Edit, View
- **ชื่อไฟล์**: สามารถคลิกเพื่อเปลี่ยนชื่อไฟล์ได้
- **พื้นที่ทำงาน**: พื้นที่สำหรับเขียนโค้ดและแสดงผลลัพธ์
- **แถบเครื่องมือ**: มีปุ่มสำหรับรันโค้ด, เพิ่มเซลล์, ลบเซลล์

### 3. ประเภทของเซลล์
Google Colab มีเซลล์ 2 ประเภทหลัก:
- **Code cell**: สำหรับเขียนและรันโค้ด Python
- **Text cell**: สำหรับเขียนข้อความบรรยายด้วย Markdown

## การใช้งาน Code Cell

Code cell ใช้สำหรับเขียนและรันโค้ด Python สามารถรันได้โดยการกดปุ่ม Play ที่อยู่ด้านซ้ายของเซลล์ หรือกด Shift+Enter

ตัวอย่างการใช้งาน Code cell:

```python
# นี่คือคอมเมนต์
print("Hello, Data Science World!")

# คำนวณตัวเลข
a = 5
b = 3
print(f"a + b = {a + b}")
```

หากต้องการติดตั้งไลบรารีที่ยังไม่มีใน Google Colab สามารถทำได้โดยใช้คำสั่ง `pip install`:

### ตัวอย่างการใช้ numpy

```python
import numpy as np

# สร้าง array ตัวเลข
arr = np.array([1, 2, 3, 4, 5])
print("Array:", arr)
print("Mean:", np.mean(arr))
```

## การใช้งาน Text Cell (Markdown)

Text cell ใช้สำหรับเขียนข้อความบรรยายด้วย Markdown สามารถเพิ่ม Text cell ได้โดยการคลิกที่ "+Text" ในแถบเครื่องมือ

### ตัวอย่างการใช้ Markdown

```markdown
# หัวข้อใหญ่
## หัวข้อรอง
### หัวข้อย่อย

ข้อความธรรมดา **ข้อความตัวหนา** *ข้อความตัวเอียง*

- รายการข้อที่ 1
- รายการข้อที่ 2
  - ข้อย่อย 2.1
  - ข้อย่อย 2.2

1. รายการลำดับที่ 1
2. รายการลำดับที่ 2

[ลิงก์ไปยัง Google](https://www.google.com)

```

### การแทรกรูปภาพ

```markdown
![ชื่อรูปภาพ](URL_ของรูปภาพ)
```

## เคล็ดลับการใช้งาน Google Colab

### คีย์ลัดที่มีประโยชน์
- **Shift+Enter**: รันเซลล์ปัจจุบันและเลื่อนไปยังเซลล์ถัดไป
- **Ctrl+Enter**: รันเซลล์ปัจจุบันโดยไม่เลื่อนไปยังเซลล์ถัดไป
- **Alt+Enter**: รันเซลล์ปัจจุบันและแทรกเซลล์ใหม่ด้านล่าง
- **Ctrl+M + B**: แทรกเซลล์ด้านล่าง
- **Ctrl+M + A**: แทรกเซลล์ด้านบน
- **Ctrl+M + D**: ลบเซลล์ปัจจุบัน

### การตรวจสอบทรัพยากร
คุณสามารถตรวจสอบทรัพยากรที่ใช้งานอยู่ได้โดยคลิกที่แท็บ "Runtime" และเลือก "Manage sessions"

### การเปลี่ยนชนิดของ Runtime
หากต้องการใช้ GPU หรือ TPU:
1. คลิกที่ "Runtime"
2. เลือก "Change runtime type"
3. เลือก "GPU" หรือ "TPU" ในส่วนของ Hardware accelerator

## กิจกรรมฝึกปฏิบัติ

### กิจกรรม 1: สร้าง Notebook แรกของคุณ
1. เข้าไปที่ [colab.research.google.com](https://colab.research.google.com)
2. สร้าง Notebook ใหม่

5. สร้าง Code cell และเขียนโค้ดง่ายๆ เช่น `print("Hello, Data Science!")`

### กิจกรรม 2: ทดลองใช้ไลบรารีพื้นฐาน
สร้าง Code cell ใหม่และทดลองใช้ไลบรารี NumPy:

```python
import numpy as np

# สร้างอาร์เรย์ตัวเลขสุ่ม
random_array = np.random.rand(5, 5)
print("Random Array:")
print(random_array)

# คำนวณค่าเฉลี่ยและส่วนเบี่ยงเบนมาตรฐาน
print(f"Mean: {np.mean(random_array)}")
print(f"Standard Deviation: {np.std(random_array)}")
```
