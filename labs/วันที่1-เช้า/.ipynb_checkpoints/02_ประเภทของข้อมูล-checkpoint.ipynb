# ประเภทของข้อมูลและแหล่งที่มา

## วัตถุประสงค์การเรียนรู้
- เข้าใจประเภทและโครงสร้างของข้อมูลแบบต่างๆ
- รู้จักแหล่งที่มาของข้อมูลทั้งภายในและภายนอกองค์กร
- เข้าใจข้อควรระวังในการเลือกใช้ข้อมูล
- สามารถเลือกใช้ข้อมูลให้เหมาะสมกับปัญหาที่ต้องการแก้ไข

## ประเภทของข้อมูล

### แบ่งตามโครงสร้าง

#### 1. ข้อมูลแบบมีโครงสร้าง (Structured Data)
- มีรูปแบบและโครงสร้างที่ชัดเจน
- จัดเก็บในรูปแบบตารางแถวและคอลัมน์
- ตัวอย่าง: ฐานข้อมูล SQL, Excel, CSV

#### 2. ข้อมูลแบบกึ่งโครงสร้าง (Semi-structured Data)
- มีโครงสร้างแต่ไม่ได้อยู่ในรูปแบบตารางแถวและคอลัมน์
- มีแท็กหรือเครื่องหมายอื่นๆ กำกับข้อมูล
- ตัวอย่าง: XML, JSON, HTML

#### 3. ข้อมูลแบบไม่มีโครงสร้าง (Unstructured Data)
- ไม่มีรูปแบบหรือโครงสร้างที่กำหนดไว้ล่วงหน้า
- ยากต่อการประมวลผลด้วยวิธีการแบบดั้งเดิม
- ตัวอย่าง: ข้อความ, รูปภาพ, วิดีโอ, ไฟล์เสียง, อีเมล

### แบ่งตามประเภทของค่าข้อมูล

#### 1. ข้อมูลเชิงปริมาณ (Quantitative Data)
- สามารถวัดค่าเป็นตัวเลขได้
- แบ่งเป็น:
  - **ข้อมูลแบบต่อเนื่อง (Continuous)**: มีค่าทศนิยมได้ เช่น น้ำหนัก, ส่วนสูง, อุณหภูมิ
  - **ข้อมูลแบบไม่ต่อเนื่อง (Discrete)**: เป็นค่าเต็ม นับได้ เช่น จำนวนลูกค้า, จำนวนสินค้า

#### 2. ข้อมูลเชิงคุณภาพ (Qualitative Data)
- บรรยายลักษณะหรือคุณสมบัติ
- แบ่งเป็น:
  - **ข้อมูลแบบระบุชื่อ (Nominal)**: ไม่มีลำดับ เช่น สี, เพศ, ประเทศ
  - **ข้อมูลแบบจัดลำดับ (Ordinal)**: มีลำดับแต่ระยะห่างไม่เท่ากัน เช่น ระดับความพึงพอใจ (น้อย, ปานกลาง, มาก)

## ตัวอย่างประเภทข้อมูลในโลกจริง

ลองนึกถึงข้อมูลร้านอาหาร:

### ข้อมูลแบบมีโครงสร้าง:
- ตารางรายการสั่งอาหาร (วันที่, รหัสลูกค้า, รายการอาหาร, จำนวน, ราคา)
- ตารางสินค้าคงคลัง (รหัสวัตถุดิบ, ชื่อวัตถุดิบ, จำนวนคงเหลือ, ราคาต้นทุน)

### ข้อมูลแบบกึ่งโครงสร้าง:
- ข้อมูล JSON จาก API การสั่งอาหารออนไลน์
- ไฟล์ XML ของการตั้งค่าเครื่อง POS

### ข้อมูลแบบไม่มีโครงสร้าง:
- รีวิวร้านอาหารจากลูกค้า
- รูปถ่ายอาหารที่ลูกค้าโพสต์ในโซเชียลมีเดีย
- วิดีโอการสอนทำอาหาร

## แหล่งที่มาของข้อมูล

### 1. แหล่งข้อมูลภายในองค์กร (Internal Data Sources)
- **ระบบธุรกรรม (Transaction Systems)**: ERP, CRM, ระบบขาย
- **ฐานข้อมูลองค์กร (Corporate Databases)**: ข้อมูลลูกค้า, ข้อมูลพนักงาน
- **เว็บไซต์และแอปพลิเคชันขององค์กร**: ข้อมูลการเข้าชม, พฤติกรรมผู้ใช้
- **อีเมลและการสื่อสารภายใน**: การโต้ตอบกับลูกค้า, ข้อมูลการสนับสนุน
- **เอกสารภายใน**: รายงาน, งานวิจัย, บันทึกการประชุม

### 2. แหล่งข้อมูลภายนอกองค์กร (External Data Sources)
- **ข้อมูลสาธารณะ (Public Data)**: ข้อมูลประชากร, ข้อมูลเศรษฐกิจ, ข้อมูลภูมิศาสตร์
- **โซเชียลมีเดีย (Social Media)**: ความคิดเห็น, แนวโน้ม, การมีส่วนร่วม
- **ข้อมูลการวิจัยตลาด**: รายงานอุตสาหกรรม, การสำรวจตลาด
- **APIs จากบริการออนไลน์**: ข้อมูลสภาพอากาศ, อัตราแลกเปลี่ยน, ราคาสินค้า
- **ข้อมูลจากพันธมิตรทางธุรกิจ**: ข้อมูลจากผู้จัดจำหน่าย, ข้อมูลจากตัวแทนจำหน่าย

## ข้อควรพิจารณาเกี่ยวกับข้อมูล

### 1. คุณภาพของข้อมูล (Data Quality)
- **ความถูกต้อง (Accuracy)**: ข้อมูลตรงกับความเป็นจริง
- **ความสมบูรณ์ (Completeness)**: ข้อมูลมีครบถ้วนไม่ขาดหาย
- **ความทันสมัย (Timeliness)**: ข้อมูลเป็นปัจจุบัน
- **ความสอดคล้อง (Consistency)**: ข้อมูลไม่ขัดแย้งกันเอง
- **ความไม่ซ้ำซ้อน (Uniqueness)**: ไม่มีข้อมูลซ้ำที่ไม่จำเป็น

### 2. ความเป็นส่วนตัวและความปลอดภัยของข้อมูล (Data Privacy & Security)
- **การปกป้องข้อมูลส่วนบุคคล**: การปฏิบัติตาม PDPA, GDPR
- **การเก็บความลับ**: การเข้ารหัสข้อมูล
- **สิทธิ์การเข้าถึง**: การจำกัดการเข้าถึงข้อมูลตามความจำเป็น

### 3. อคติในข้อมูล (Data Bias)
- **อคติจากการเลือกตัวอย่าง (Sampling Bias)**: กลุ่มตัวอย่างไม่เป็นตัวแทนที่ดีของประชากร
- **อคติจากการวัด (Measurement Bias)**: วิธีการวัดหรือเครื่องมือมีอคติ
- **อคติจากการตีความ (Interpretation Bias)**: การตีความข้อมูลผิดพลาด

## หลักการเลือกใช้ข้อมูล

### 1. คำนึงถึงวัตถุประสงค์ (Purpose-driven)
- เลือกข้อมูลที่เกี่ยวข้องกับปัญหาหรือคำถามที่ต้องการตอบ
- ระบุตัวชี้วัดที่เหมาะสม (KPIs) สำหรับเป้าหมายของคุณ

### 2. พิจารณาความพร้อมใช้งานของข้อมูล (Data Availability)
- ตรวจสอบว่ามีข้อมูลที่ต้องการหรือไม่
- ประเมินต้นทุนและเวลาในการรวบรวมข้อมูล

### 3. ประเมินความน่าเชื่อถือของแหล่งข้อมูล (Source Reliability)
- พิจารณาความน่าเชื่อถือของแหล่งที่มา
- ตรวจสอบวิธีการเก็บข้อมูล

### 4. หลีกเลี่ยงการจมอยู่กับข้อมูลมากเกินไป (Avoid Data Overload)
- เริ่มต้นด้วยชุดข้อมูลที่จำเป็นจริงๆ
- ขยายขอบเขตเมื่อจำเป็น

## กิจกรรมฝึกปฏิบัติ

### กิจกรรม 1: ระบุประเภทของข้อมูล
ให้ระบุว่าข้อมูลต่อไปนี้เป็นประเภทใด (มีโครงสร้าง/กึ่งโครงสร้าง/ไม่มีโครงสร้าง และ เชิงปริมาณ/เชิงคุณภาพ)

1. รายการคำสั่งซื้อในระบบ e-commerce
2. รีวิวสินค้าที่เขียนโดยลูกค้า
3. ไฟล์ JSON จาก API โซเชียลมีเดีย
4. ระดับความพึงพอใจของลูกค้า (1-5 ดาว)
5. รูปภาพสินค้า
6. อายุของลูกค้า
7. ประเภทสินค้า (เสื้อผ้า, อาหาร, อิเล็กทรอนิกส์)

### กิจกรรม 2: ระบุแหล่งที่มาของข้อมูล
สมมติว่าคุณเป็นนักวิเคราะห์ข้อมูลในร้านค้าปลีก และต้องการวิเคราะห์เพื่อเพิ่มยอดขาย ให้ระบุแหล่งข้อมูลที่คุณจะใช้ ทั้งภายในและภายนอกองค์กร

## สรุป

- ข้อมูลสามารถแบ่งได้หลายประเภท เช่น แบ่งตามโครงสร้าง (มีโครงสร้าง/กึ่งโครงสร้าง/ไม่มีโครงสร้าง) หรือแบ่งตามประเภทค่า (เชิงปริมาณ/เชิงคุณภาพ)
- แหล่งข้อมูลมีทั้งจากภายในองค์กร (เช่น ระบบ ERP, ฐานข้อมูลลูกค้า) และภายนอกองค์กร (เช่น ข้อมูลสาธารณะ, โซเชียลมีเดีย)
- การเลือกใช้ข้อมูลควรคำนึงถึงคุณภาพของข้อมูล ความเป็นส่วนตัว และอคติที่อาจเกิดขึ้น
- การเลือกข้อมูลที่เหมาะสมกับปัญหาจะช่วยให้การวิเคราะห์มีประสิทธิภาพและได้ผลลัพธ์ที่เชื่อถือได้
