# การสร้าง Visualization ด้วย BI Tools

## วัตถุประสงค์การเรียนรู้
- เข้าใจความสำคัญของเครื่องมือ Business Intelligence (BI Tools)
- รู้จักและสามารถใช้ Power BI ในการสร้างการแสดงผลข้อมูล
- สามารถนำเข้าข้อมูลและเชื่อมต่อกับแหล่งข้อมูลต่างๆ
- สร้าง Dashboard และ Report ที่มีประสิทธิภาพ
- ออกแบบการแสดงผลข้อมูลให้เหมาะสมกับกลุ่มเป้าหมาย

## 1. ความสำคัญของเครื่องมือ Business Intelligence (BI Tools)

เครื่องมือ Business Intelligence (BI Tools) เป็นซอฟต์แวร์ที่ออกแบบมาเพื่อรวบรวม วิเคราะห์ และแสดงผลข้อมูลทางธุรกิจในรูปแบบที่เข้าใจง่าย ซึ่งมีความสำคัญหลายประการ:

1. **รวมข้อมูลจากหลายแหล่ง**: สามารถดึงข้อมูลจากหลายแหล่งและหลายรูปแบบมาไว้ในที่เดียวกัน
2. **สร้างการแสดงผลข้อมูลแบบโต้ตอบ**: ช่วยให้ผู้ใช้สามารถสำรวจข้อมูลและค้นหาข้อมูลเชิงลึกได้อย่างมีประสิทธิภาพ
3. **แบ่งปันและทำงานร่วมกัน**: ช่วยให้ทีมสามารถแบ่งปันและทำงานร่วมกันบนข้อมูลและการวิเคราะห์เดียวกัน
4. **ติดตามตัวชี้วัดสำคัญ (KPIs)**: ช่วยให้องค์กรสามารถติดตามความคืบหน้าและความสำเร็จของเป้าหมายได้
5. **ตัดสินใจโดยอิงจากข้อมูล**: ช่วยให้ผู้บริหารและพนักงานสามารถตัดสินใจได้อย่างมีประสิทธิภาพโดยอิงจากข้อมูลที่เป็นปัจจุบัน

ในบทเรียนนี้ เราจะเน้นการใช้ Power BI ซึ่งเป็นหนึ่งในเครื่องมือ BI ที่ได้รับความนิยมมากที่สุดในปัจจุบัน

## 2. แนะนำ Power BI และเครื่องมือ BI อื่นๆ

เครื่องมือ Business Intelligence (BI) มีหลากหลายตัวในท้องตลาด แต่ละตัวมีจุดเด่นและข้อจำกัดแตกต่างกันไป ในบทเรียนนี้เราจะเน้นที่ Power BI แต่จะแนะนำเครื่องมืออื่นๆ เพื่อให้เข้าใจภาพรวมของเครื่องมือ BI ในปัจจุบัน

### 2.1 Power BI

**Power BI** เป็นเครื่องมือวิเคราะห์ข้อมูลและสร้างการแสดงผลข้อมูลที่พัฒนาโดยไมโครซอฟท์ ประกอบด้วย:

- **Power BI Desktop**: แอปพลิเคชันสำหรับสร้างรายงานและแดชบอร์ด
- **Power BI Service**: บริการคลาวด์สำหรับแบ่งปันและเผยแพร่รายงาน
- **Power BI Mobile**: แอปพลิเคชันมือถือสำหรับดูรายงานและแดชบอร์ด

**จุดเด่นของ Power BI:**
- การเชื่อมต่อกับแหล่งข้อมูลที่หลากหลาย
- DAX (Data Analysis Expressions) สำหรับการคำนวณที่ซับซ้อน
- การสร้างการแสดงผลข้อมูลแบบโต้ตอบ
- การรวมกับผลิตภัณฑ์อื่นๆ ของไมโครซอฟท์
- การอัปเดตและพัฒนาอย่างต่อเนื่อง

### 2.2 เครื่องมือ BI อื่นๆ ในตลาด

**1. Tableau**
- จุดเด่น: การสร้างการแสดงผลข้อมูลที่สวยงาม, การใช้งานที่ง่าย
- ข้อจำกัด: ราคาค่อนข้างสูง

**2. QlikView/Qlik Sense**
- จุดเด่น: การวิเคราะห์ข้อมูลแบบ In-Memory, การค้นหาข้อมูลที่รวดเร็ว
- ข้อจำกัด: เส้นการเรียนรู้ค่อนข้างชัน

**3. Looker (Google)**
- จุดเด่น: การรวมกับ Google Cloud, การสร้างโมเดลข้อมูลที่แข็งแกร่ง
- ข้อจำกัด: ต้องมีทักษะการเขียน LookML

**4. SAP BusinessObjects**
- จุดเด่น: การรวมกับระบบ SAP, การรองรับองค์กรขนาดใหญ่
- ข้อจำกัด: ความซับซ้อนในการติดตั้งและดูแลรักษา

**5. เครื่องมือโอเพนซอร์ส (Open Source)**
- ตัวอย่าง: Metabase, Redash, Apache Superset
- จุดเด่น: ไม่มีค่าใช้จ่าย, สามารถปรับแต่งได้ตามต้องการ
- ข้อจำกัด: อาจต้องมีทักษะทางเทคนิคในการติดตั้งและดูแลรักษา

## 3. การติดตั้งและใช้งาน Power BI Desktop

ก่อนที่เราจะเริ่มสร้างการแสดงผลข้อมูลด้วย Power BI เราจำเป็นต้องติดตั้งและทำความเข้าใจพื้นฐานของ Power BI Desktop

### 3.1 การติดตั้ง Power BI Desktop

1. **ดาวน์โหลด Power BI Desktop**
   - ไปที่เว็บไซต์ [https://powerbi.microsoft.com/desktop/](https://powerbi.microsoft.com/desktop/)
   - คลิกที่ "ดาวน์โหลดฟรี"
   - เลือกเวอร์ชันที่เหมาะสมกับระบบปฏิบัติการของคุณ

2. **ขั้นตอนการติดตั้ง**
   - เปิดไฟล์ติดตั้งที่ดาวน์โหลดมา
   - ทำตามคำแนะนำในการติดตั้ง
   - รอให้การติดตั้งเสร็จสิ้น

3. **เริ่มต้นใช้งาน**
   - เปิด Power BI Desktop
   - ลงชื่อเข้าใช้ด้วยบัญชี Microsoft (หากมี)

### 3.2 ส่วนประกอบหลักของ Power BI Desktop

**1. ส่วนรายงาน (Report View)**
- พื้นที่สำหรับสร้างและออกแบบการแสดงผลข้อมูล
- สามารถเพิ่มและจัดเรียงวิชวลต่างๆ

**2. ส่วนข้อมูล (Data View)**
- แสดงข้อมูลในรูปแบบตาราง
- ตรวจสอบประเภทข้อมูลและค่าต่างๆ

**3. ส่วนความสัมพันธ์ (Relationship View)**
- แสดงความสัมพันธ์ระหว่างตารางข้อมูล
- สร้างและแก้ไขความสัมพันธ์

**4. แผงวิชวล (Visualizations Pane)**
- เลือกประเภทของการแสดงผลข้อมูล
- ปรับแต่งรูปแบบและสีของวิชวล

**5. แผงฟิลด์ (Fields Pane)**
- แสดงตารางและฟิลด์ที่มีอยู่
- ลากฟิลด์ไปยังพื้นที่สร้างวิชวล

**6. ริบบอน (Ribbon)**
- เครื่องมือสำหรับการจัดการรายงาน
- ตัวเลือกการนำเข้าข้อมูล แปลงข้อมูล และการเผยแพร่

### 3.3 การใช้งานเบื้องต้น

**การเปิดไฟล์ Power BI**
```
File > Open > Browse to your .pbix file
```

**การสร้างไฟล์ใหม่**
```
File > New
```

**การบันทึกไฟล์**
```
File > Save / Save As
```

**การนำเข้าข้อมูล**
```
Home > Get Data
```

**การเผยแพร่รายงาน**
```
Home > Publish
```

## 4. การนำเข้าข้อมูลและการเชื่อมต่อกับแหล่งข้อมูล

การนำเข้าข้อมูลเป็นขั้นตอนแรกและสำคัญในการสร้างการแสดงผลข้อมูลด้วย Power BI ซึ่ง Power BI สามารถเชื่อมต่อกับแหล่งข้อมูลได้หลากหลาย

### 4.1 แหล่งข้อมูลที่ Power BI รองรับ

Power BI สามารถเชื่อมต่อกับแหล่งข้อมูลได้หลากหลาย ทั้งจากไฟล์ ฐานข้อมูล บริการคลาวด์ และอื่นๆ

**1. ไฟล์**
- Excel (.xlsx, .xls)
- CSV (.csv)
- Text/ข้อความ (.txt)
- XML (.xml)
- JSON (.json)
- PDF (.pdf)

**2. ฐานข้อมูล**
- SQL Server
- MySQL
- PostgreSQL
- Oracle
- IBM DB2
- Amazon Redshift
- Snowflake
- SAP HANA

**3. บริการคลาวด์**
- Azure SQL Database
- Azure Synapse Analytics
- Amazon Redshift
- Google BigQuery
- Salesforce
- Dynamics 365

**4. บริการออนไลน์**
- SharePoint
- Dynamics 365
- Google Analytics
- Facebook
- GitHub

**5. อื่นๆ**
- Web (HTML)
- ODBC
- OLE DB
- R Script
- Python Script

### 4.2 ขั้นตอนการนำเข้าข้อมูล

การนำเข้าข้อมูลใน Power BI มีขั้นตอนดังนี้:

**1. เริ่มนำเข้าข้อมูล**
- คลิกที่ "Get Data" ในแท็บ Home
- เลือกแหล่งข้อมูลที่ต้องการ

**2. เลือกข้อมูล**
- เลือกไฟล์หรือป้อนข้อมูลการเชื่อมต่อ
- เลือกตารางหรือชีทที่ต้องการ
- คลิก "Load" หรือ "Transform Data"

**3. ปรับแต่งข้อมูล (ถ้าจำเป็น)**
- ใช้ Power Query Editor เพื่อปรับแต่งข้อมูล
- เปลี่ยนประเภทข้อมูล
- กรองหรือจัดเรียงข้อมูล
- สร้างคอลัมน์ใหม่
- รวมหรือแยกตาราง

**4. โหลดข้อมูล**
- คลิก "Close & Apply" เพื่อโหลดข้อมูลเข้าสู่โมเดล

### 4.3 การสร้างความสัมพันธ์ระหว่างตาราง

เมื่อมีหลายตารางในโมเดลข้อมูล การสร้างความสัมพันธ์ระหว่างตารางจะช่วยให้สามารถวิเคราะห์ข้อมูลข้ามตารางได้

**ขั้นตอนการสร้างความสัมพันธ์**

1. **สลับไปยังมุมมองความสัมพันธ์**
   - คลิกที่ไอคอน "Relationships" ทางด้านซ้าย

2. **สร้างความสัมพันธ์ใหม่**
   - คลิกที่ "Manage Relationships" ในแท็บ Home
   - คลิก "New"
   - เลือกตารางและคอลัมน์ที่ต้องการเชื่อมโยง
   - กำหนดประเภทความสัมพันธ์ (Many-to-One, One-to-One, Many-to-Many)

3. **แก้ไขความสัมพันธ์**
   - ดับเบิลคลิกที่เส้นความสัมพันธ์
   - แก้ไขการตั้งค่าตามต้องการ

**ประเภทของความสัมพันธ์**
- **One-to-Many (1:N)**: ความสัมพันธ์ทั่วไป เช่น หนึ่งลูกค้ามีหลายคำสั่งซื้อ
- **One-to-One (1:1)**: ข้อมูลมีความสัมพันธ์แบบหนึ่งต่อหนึ่ง
- **Many-to-Many (M:N)**: ต้องมีตารางเชื่อมโยงอยู่ตรงกลาง

### 4.4 การรีเฟรชข้อมูล

การรีเฟรชข้อมูลเป็นกระบวนการอัปเดตข้อมูลในรายงาน Power BI เมื่อข้อมูลต้นทางมีการเปลี่ยนแปลง

**การรีเฟรชข้อมูลแบบแมนนวล**
- คลิกที่ "Refresh" ในแท็บ Home
- เลือก "Refresh" เพื่อรีเฟรชทุกแหล่งข้อมูล
- หรือเลือก "Refresh [ชื่อแหล่งข้อมูล]" เพื่อรีเฟรชเฉพาะแหล่งข้อมูลนั้น

**การตั้งค่าการรีเฟรชอัตโนมัติ (ใน Power BI Service)**
- ไปที่ Dataset Settings
- ตั้งค่า Scheduled Refresh
- กำหนดความถี่และเวลาในการรีเฟรช

**ข้อควรระวัง**
- การรีเฟรชข้อมูลขนาดใหญ่อาจใช้เวลานาน
- ตรวจสอบข้อจำกัดของการรีเฟรชตามแผนการใช้งาน Power BI
- อาจต้องตั้งค่า Gateway สำหรับข้อมูลภายในองค์กร

## 5. การสร้าง Dashboard และ Report

การสร้าง Dashboard และ Report เป็นขั้นตอนสำคัญในการนำเสนอข้อมูลอย่างมีประสิทธิภาพ

### 5.1 การสร้างวิชวลพื้นฐาน

Power BI มีวิชวลหลากหลายรูปแบบให้เลือกใช้ตามความเหมาะสมกับข้อมูลและวัตถุประสงค์

**ขั้นตอนการสร้างวิชวล**
1. เลือกพื้นที่ว่างในรายงาน
2. เลือกประเภทวิชวลจากแผง Visualizations
3. ลากฟิลด์ที่ต้องการใช้ไปยังกล่อง Fields

**ประเภทของวิชวลพื้นฐาน**

**1. กราฟแท่ง (Bar/Column Chart)**
- เหมาะสำหรับการเปรียบเทียบระหว่างหมวดหมู่
- วิธีใช้: ลากฟิลด์หมวดหมู่ไปที่ Axis และฟิลด์ค่าไปที่ Values

**2. กราฟเส้น (Line Chart)**
- เหมาะสำหรับแสดงแนวโน้มตามเวลา
- วิธีใช้: ลากฟิลด์เวลาไปที่ Axis และฟิลด์ค่าไปที่ Values

**3. กราฟวงกลม (Pie Chart)**
- เหมาะสำหรับแสดงสัดส่วน
- วิธีใช้: ลากฟิลด์หมวดหมู่ไปที่ Legend และฟิลด์ค่าไปที่ Values

**4. การ์ด (Card/KPI)**
- เหมาะสำหรับแสดงตัวเลขสำคัญ
- วิธีใช้: ลากฟิลด์ค่าไปที่ Fields

**5. ตาราง/เมทริกซ์ (Table/Matrix)**
- เหมาะสำหรับแสดงข้อมูลรายละเอียด
- วิธีใช้: ลากฟิลด์ที่ต้องการแสดงไปที่ Values/Columns/Rows

**6. แผนที่ (Map)**
- เหมาะสำหรับข้อมูลที่มีตำแหน่งภูมิศาสตร์
- วิธีใช้: ลากฟิลด์ตำแหน่งไปที่ Location และฟิลด์ค่าไปที่ Size/Color

### 5.2 การสร้างแดชบอร์ดที่มีประสิทธิภาพ

แดชบอร์ดที่มีประสิทธิภาพควรนำเสนอข้อมูลสำคัญอย่างชัดเจนและเข้าใจง่าย ต่อไปนี้เป็นหลักการสำคัญในการสร้างแดชบอร์ด:

**1. กำหนดวัตถุประสงค์ให้ชัดเจน**
- ระบุกลุ่มผู้ใช้งานและความต้องการ
- กำหนดคำถามที่แดชบอร์ดต้องตอบได้
- เลือกตัวชี้วัดสำคัญ (KPIs) ที่เหมาะสม

**2. จัดวางองค์ประกอบอย่างมีประสิทธิภาพ**
- ใช้หลัก Z-pattern หรือ F-pattern ในการจัดวาง
- วางข้อมูลสำคัญไว้ด้านบนซ้าย
- จัดกลุ่มวิชวลที่เกี่ยวข้องกันไว้ด้วยกัน
- ใช้พื้นที่อย่างคุ้มค่า ไม่แน่นหรือโล่งเกินไป

**3. ออกแบบให้ดูง่ายและเข้าใจง่าย**
- ใช้สีอย่างมีความหมาย
- เลือกฟอนต์ที่อ่านง่าย
- มีระยะห่างที่เหมาะสม
- ใช้เส้นแบ่งหรือพื้นหลังเพื่อแบ่งพื้นที่

**4. เพิ่มความสามารถในการโต้ตอบ**
- ใช้ Slicers สำหรับกรองข้อมูล
- ตั้งค่า Drill Through สำหรับดูรายละเอียด
- เชื่อมโยงวิชวลเข้าด้วยกัน (Cross-filtering)

**5. ทำให้เรียบง่ายแต่มีประสิทธิภาพ**
- หลีกเลี่ยงการใช้วิชวลมากเกินไป
- ลดองค์ประกอบที่ไม่จำเป็น
- มุ่งเน้นที่ข้อความหลักที่ต้องการสื่อสาร

### 5.3 การเพิ่มความสามารถในการโต้ตอบ

การสร้างความสามารถในการโต้ตอบช่วยให้ผู้ใช้สามารถสำรวจข้อมูลและค้นหาข้อมูลเชิงลึกได้ด้วยตนเอง

**1. ตัวกรอง (Slicers)**
- วิธีเพิ่ม: เลือก Slicer จากแผง Visualizations แล้วลากฟิลด์ที่ต้องการกรอง
- ประโยชน์: ช่วยให้ผู้ใช้สามารถกรองข้อมูลในทุกวิชวลพร้อมกัน
- ตัวอย่าง: กรองตามช่วงเวลา ภูมิภาค หรือผลิตภัณฑ์

**2. การเจาะลึก (Drill Down)**
- วิธีตั้งค่า: จัดลำดับชั้นใน Fields เช่น ปี > ไตรมาส > เดือน
- การใช้งาน: คลิกที่จุดข้อมูลในวิชวลแล้วใช้เครื่องมือ Drill Down
- ประโยชน์: ช่วยให้ผู้ใช้สามารถดูข้อมูลในระดับที่ละเอียดขึ้น

**3. การเชื่อมโยงระหว่างหน้า (Drill Through)**
- วิธีตั้งค่า: สร้างหน้ารายละเอียด แล้วตั้งค่า Drill Through Fields
- การใช้งาน: คลิกขวาที่จุดข้อมูลแล้วเลือก Drill Through
- ประโยชน์: ช่วยให้ผู้ใช้สามารถดูรายละเอียดเพิ่มเติมได้

**4. เคล็ดลับ (Tooltips)**
- วิธีตั้งค่า: ปรับแต่งแท็บ Tooltip ในเมนู Format Visual
- การใช้งาน: ผู้ใช้จะเห็นเคล็ดลับเมื่อวางเมาส์เหนือจุดข้อมูล
- ประโยชน์: แสดงข้อมูลเพิ่มเติมโดยไม่ต้องใช้พื้นที่ในรายงาน

**5. Bookmarks และ Buttons**
- วิธีตั้งค่า: สร้าง Bookmark แล้วเชื่อมโยงกับ Button
- การใช้งาน: ผู้ใช้คลิกที่ปุ่มเพื่อเปลี่ยนมุมมองหรือการตั้งค่า
- ประโยชน์: ช่วยให้ผู้ใช้สามารถสลับระหว่างมุมมองต่างๆ ได้

### 5.4 การเผยแพร่รายงานและแดชบอร์ด

เมื่อสร้างรายงานและแดชบอร์ดเสร็จแล้ว ขั้นตอนต่อไปคือการเผยแพร่เพื่อให้ผู้ใช้อื่นๆ สามารถเข้าถึงได้

**1. การเผยแพร่ไปยัง Power BI Service**
- คลิกที่ "Publish" ในแท็บ Home
- เลือก Workspace ที่ต้องการเผยแพร่
- รอให้การเผยแพร่เสร็จสิ้น

**2. การแบ่งปันรายงาน**
- ในเว็บไซต์ Power BI เปิดรายงานที่ต้องการแบ่งปัน
- คลิกที่ "Share" หรือ "Access"
- ป้อนอีเมลหรือกลุ่มผู้ใช้ที่ต้องการแบ่งปัน
- เลือกสิทธิ์การเข้าถึง (ดู, แก้ไข)

**3. การฝัง (Embed) รายงาน**
- ในเว็บไซต์ Power BI เปิดรายงานที่ต้องการฝัง
- คลิกที่ "File > Embed" หรือ "Share > Embed"
- คัดลอกโค้ดและนำไปใช้ในเว็บไซต์หรือแอปพลิเคชัน

**4. การส่งออกเป็นไฟล์**
- ในเว็บไซต์ Power BI เปิดรายงานที่ต้องการส่งออก
- คลิกที่ "File > Export to"
- เลือกรูปแบบไฟล์ที่ต้องการ (PDF, PowerPoint, etc.)

**5. การตั้งค่าการเข้าถึง**
- การตั้งค่า Row-Level Security (RLS) เพื่อจำกัดการเข้าถึงข้อมูล
- การตั้งค่า Workspace Access เพื่อควบคุมการเข้าถึงทรัพยากร
- การใช้ Power BI Apps เพื่อแพ็กเกจรายงานสำหรับผู้ใช้ปลายทาง

## 6. การประยุกต์ใช้ Power BI ในงานจริง

เรามาดูตัวอย่างการประยุกต์ใช้ Power BI ในงานจริงในหลากหลายอุตสาหกรรม

### 6.1 กรณีศึกษาและตัวอย่างการใช้งาน

**1. การขายและการตลาด**
- **การวิเคราะห์ยอดขาย**: ติดตามยอดขายตามภูมิภาค, ผลิตภัณฑ์, และช่องทางการขาย
- **การวิเคราะห์ลูกค้า**: วิเคราะห์พฤติกรรมลูกค้า, ความพึงพอใจ, และการรักษาลูกค้า
- **การวิเคราะห์แคมเปญ**: วัดประสิทธิภาพของแคมเปญการตลาดและ ROI

**2. การเงินและการบัญชี**
- **การวิเคราะห์งบการเงิน**: แสดงมุมมองงบกำไรขาดทุน, งบแสดงฐานะการเงินแบบโต้ตอบ
- **การติดตามงบประมาณ**: เปรียบเทียบงบประมาณกับค่าใช้จ่ายจริง
- **การพยากรณ์กระแสเงินสด**: สร้างแบบจำลองและคาดการณ์กระแสเงินสด

**3. การผลิตและห่วงโซ่อุปทาน**
- **การติดตามประสิทธิภาพการผลิต**: ติดตาม KPIs การผลิต เช่น อัตราของเสีย, ประสิทธิภาพโดยรวม
- **การจัดการสินค้าคงคลัง**: ตรวจสอบระดับสินค้าคงคลัง, รอบเวลาการหมุนเวียน, และการคาดการณ์
- **การติดตามซัพพลายเออร์**: วิเคราะห์ประสิทธิภาพซัพพลายเออร์, เวลานำส่ง, และคุณภาพ

**4. ทรัพยากรบุคคล**
- **การวิเคราะห์บุคลากร**: ตรวจสอบอัตราการลาออก, การจ้างงาน, และประสิทธิภาพ
- **การติดตามการฝึกอบรม**: วัดประสิทธิภาพโปรแกรมการฝึกอบรมและทักษะพนักงาน
- **การวิเคราะห์ค่าตอบแทน**: เปรียบเทียบค่าตอบแทนกับผลการปฏิบัติงานและตลาด

**5. สาธารณสุขและการดูแลสุขภาพ**
- **การติดตามผู้ป่วย**: ติดตามการรับเข้า, การจำหน่าย, และตัวชี้วัดผลลัพธ์
- **การวิเคราะห์ผลลัพธ์ทางคลินิก**: เปรียบเทียบวิธีการรักษาและผลลัพธ์
- **การจัดการทรัพยากร**: ติดตามการใช้ทรัพยากร เช่น เตียง, อุปกรณ์, และบุคลากร

### 6.2 แนวทางปฏิบัติที่ดีในการใช้ Power BI

เมื่อใช้งาน Power BI มีแนวทางปฏิบัติที่ดีเพื่อให้ได้ผลลัพธ์ที่มีประสิทธิภาพ ดังนี้:

**1. การวางแผนและออกแบบ**
- กำหนดวัตถุประสงค์และกลุ่มเป้าหมายให้ชัดเจน
- ออกแบบโมเดลข้อมูล (Data Model) ที่มีประสิทธิภาพ
- จัดโครงสร้างรายงานให้เป็นหมวดหมู่และเป็นระบบ

**2. การเตรียมและจัดการข้อมูล**
- ทำความสะอาดข้อมูลก่อนนำเข้า (Data Cleaning)
- ใช้ Power Query Editor ปรับแต่งข้อมูลให้พร้อมใช้งาน
- ใช้ DAX (Data Analysis Expressions) สร้างการคำนวณที่ซับซ้อน

**3. การออกแบบวิชวล**
- เลือกวิชวลที่เหมาะสมกับประเภทข้อมูล
- ใช้สีอย่างมีความหมายและสอดคล้องกัน
- จัดองค์ประกอบให้อ่านง่ายและเป็นระเบียบ

**4. การเพิ่มประสิทธิภาพ**
- ลดขนาดไฟล์ให้เล็กที่สุดเท่าที่เป็นไปได้
- ใช้ Calculated Columns เมื่อจำเป็นเท่านั้น
- ตรวจสอบและปรับแต่งข้อมูลเชิงสัมพันธ์

**5. การทดสอบและปรับปรุง**
- ทดสอบกับผู้ใช้จริงและรับฟังความคิดเห็น
- ตรวจสอบความถูกต้องของข้อมูลอย่างสม่ำเสมอ
- ปรับปรุงและอัปเดตรายงานตามความต้องการที่เปลี่ยนแปลง

## 7. แบบฝึกหัด: สร้างแดชบอร์ดการวิเคราะห์ยอดขาย

ลองฝึกการสร้างแดชบอร์ดการวิเคราะห์ยอดขายอย่างง่ายด้วย Power BI โดยใช้ขั้นตอนต่อไปนี้:

### 7.1 ข้อมูลสำหรับแบบฝึกหัด

1. **นำเข้าข้อมูลตัวอย่าง**
   - ใช้ไฟล์ข้อมูลยอดขาย (ตัวอย่างเช่น SalesData.xlsx)
   - หรือใช้ข้อมูลตัวอย่างของ Power BI (Sample Data)

2. **ข้อมูลที่ควรมี**
   - วันที่ขาย (ปี, ไตรมาส, เดือน)
   - ผลิตภัณฑ์ (ชื่อ, หมวดหมู่, ราคา)
   - ลูกค้า (ชื่อ, ภูมิภาค, เมือง)
   - ยอดขาย (จำนวน, มูลค่า, กำไร)

### 7.2 ขั้นตอนการสร้างแดชบอร์ด

1. **นำเข้าและเตรียมข้อมูล**
   - เปิด Power BI Desktop
   - ไปที่ 'Home > Get Data' และเลือกแหล่งข้อมูล
   - ใช้ Power Query Editor ปรับแต่งข้อมูล เช่น เปลี่ยนประเภทข้อมูล, สร้างคอลัมน์เพิ่มเติม

2. **สร้างความสัมพันธ์ระหว่างตาราง (ถ้ามีหลายตาราง)**
   - ไปที่มุมมอง 'Relationships'
   - สร้างความสัมพันธ์ระหว่างตารางตามคอลัมน์ที่เกี่ยวข้อง

3. **สร้างวิชวลพื้นฐาน**
   - กราฟแท่งแสดงยอดขายตามผลิตภัณฑ์
   - กราฟเส้นแสดงแนวโน้มยอดขายตามเวลา
   - แผนที่แสดงยอดขายตามภูมิภาค
   - การ์ดแสดงตัวเลขยอดขายรวม, กำไรรวม

4. **เพิ่มการโต้ตอบ**
   - เพิ่ม Slicers สำหรับกรองตามปี, หมวดหมู่ผลิตภัณฑ์, ภูมิภาค
   - ตั้งค่า Cross-filtering ระหว่างวิชวล
   - เพิ่ม Drill-down จากปี > ไตรมาส > เดือน

5. **ออกแบบและจัดวางองค์ประกอบ**
   - จัดวางองค์ประกอบให้เป็นระเบียบ
   - ใช้สีที่สอดคล้องกัน
   - เพิ่มหัวข้อและคำอธิบาย

### 7.3 ข้อคำถามและแนวทางวิเคราะห์

เมื่อสร้างแดชบอร์ดเสร็จแล้ว ลองตอบคำถามต่อไปนี้:

1. ผลิตภัณฑ์ใดมียอดขายสูงสุด? ผลิตภัณฑ์ใดมีกำไรสูงสุด?
2. แนวโน้มยอดขายในช่วงเวลาที่ผ่านมาเป็นอย่างไร?
3. ภูมิภาคใดมียอดขายสูงสุดและต่ำสุด?
4. มีความสัมพันธ์ระหว่างราคาผลิตภัณฑ์กับปริมาณการขายหรือไม่?
5. ช่วงเวลาใดในรอบปีที่มียอดขายสูงสุด? มีความแตกต่างตามภูมิภาคหรือไม่?

## 8. สรุป

การสร้าง Visualization ด้วย BI Tools โดยเฉพาะ Power BI เป็นทักษะสำคัญในปัจจุบันที่ช่วยให้องค์กรสามารถวิเคราะห์ข้อมูลและตัดสินใจได้อย่างมีประสิทธิภาพ

ในบทเรียนนี้ เราได้เรียนรู้:

1. **ความสำคัญของเครื่องมือ Business Intelligence (BI Tools)** ในการรวบรวม วิเคราะห์ และแสดงผลข้อมูลทางธุรกิจ
2. **Power BI และเครื่องมือ BI อื่นๆ** ที่มีในตลาด พร้อมจุดเด่นและข้อจำกัดของแต่ละเครื่องมือ
3. **การติดตั้งและใช้งาน Power BI Desktop** ซึ่งเป็นเครื่องมือพื้นฐานในการสร้างรายงาน
4. **การนำเข้าข้อมูลและการเชื่อมต่อกับแหล่งข้อมูล** จากหลากหลายแหล่ง
5. **การสร้าง Dashboard และ Report** ที่มีประสิทธิภาพและตอบโจทย์ความต้องการของผู้ใช้
6. **การเพิ่มความสามารถในการโต้ตอบ** เพื่อให้ผู้ใช้สามารถสำรวจข้อมูลได้ด้วยตนเอง
7. **การเผยแพร่รายงาน** เพื่อให้ผู้ใช้อื่นๆ สามารถเข้าถึงได้
8. **การประยุกต์ใช้ในงานจริง** ในหลากหลายอุตสาหกรรมและหน่วยงาน

ทักษะการสร้าง Visualization ด้วย BI Tools นี้จะเป็นประโยชน์อย่างมากในการนำเสนอข้อมูลและช่วยในการตัดสินใจ ไม่ว่าจะในองค์กรขนาดเล็กหรือขนาดใหญ่
